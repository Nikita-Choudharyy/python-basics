# 🔍 Regular Expressions (Regex) in Python

## 📖 Introduction

A **Regular Expression (Regex)** is a sequence of characters that defines a search pattern.

Regex is used to search, match, extract, validate, and replace text based on specific patterns.

Python provides built-in support for Regular Expressions through the **`re`** module.

# 📦 The `re` Module

Python provides the **`re`** module to work with Regular Expressions.

Before using Regex functions, the module must be imported.

````python
import re
````

### Explanation

The `re` module contains functions for searching, matching, splitting, replacing, and extracting text using Regular Expressions.

# 🔍 re.search()

The `search()` function searches the string for the **first occurrence** of a pattern.

If a match is found, it returns a **Match object**. Otherwise, it returns `None`.

````python
import re

text = "I am learning Python."

result = re.search("Python", text)

print(result)
````

### Explanation

The word **"Python"** exists in the string, so `search()` returns a Match object containing information about the match.

````python
text = "Python"

print(re.search("Java", text))
````

### Explanation

Since `"Java"` is not present in the string, `search()` returns `None`.

# 🎯 re.match()

The `match()` function checks whether the pattern matches **only at the beginning** of the string.

````python
text = "Python Programming"

print(re.match("Python", text))
````

### Explanation

Since the string starts with `"Python"`, the pattern matches successfully.

````python
text = "Learning Python"

print(re.match("Python", text))
````

### Explanation

Although `"Python"` exists in the string, it does not appear at the beginning.

Therefore, `match()` returns `None`.

# 📋 re.findall()

The `findall()` function returns **all occurrences** of a pattern as a list.

````python
text = "Python Java Python SQL Python"

print(re.findall("Python", text))
````

### Explanation

The `findall()` method searches the entire string and returns every occurrence of `"Python"` as a list.

# 🔄 re.sub()

The `sub()` function replaces all occurrences of a pattern with another string.

````python
text = "I love Java"

print(re.sub("Java", "Python", text))
````

### Explanation

The `sub()` function replaces every occurrence of `"Java"` with `"Python"` and returns a new string.

## 🌍 Real-World Example: Email Validation

````python
email = "nikita@gmail.com"

pattern = r'^[\w\.-]+@[\w\.-]+\.\w+$'

if re.match(pattern, email):
    print("Valid Email")
else:
    print("Invalid Email")
````

### Explanation

The regular expression checks whether the email follows a basic email format.

If the pattern matches, the email is considered valid.

# 🔣 Common Regular Expression Symbols

Regular Expressions use special symbols, called **metacharacters**, to define search patterns.

These symbols help match digits, letters, spaces, and many other text patterns efficiently.

## 1️⃣ `\d` — Match Digits

````python
text = "My PIN is 4589"

print(re.findall(r"\d", text))
````

### Explanation

The `\d` pattern matches every digit (`0–9`) found in the string.

## 2️⃣ `\D` — Match Non-Digits

````python
text = "Python123"

print(re.findall(r"\D", text))
````

### Explanation

The `\D` pattern matches every character that is **not a digit**.

## 3️⃣ `\w` — Match Word Characters

````python
text = "Python_123"

print(re.findall(r"\w", text))
````

### Explanation

The `\w` pattern matches:

- Letters (`A-Z`, `a-z`)
- Digits (`0-9`)
- Underscore (`_`)

## 4️⃣ `\W` — Match Non-Word Characters

````python
text = "Python@123!"

print(re.findall(r"\W", text))
````

### Explanation

The `\W` pattern matches characters that are **not letters, digits, or underscores**.

## 5️⃣ `\s` — Match Whitespace

````python
text = "Python Programming"

print(re.findall(r"\s", text))
````

### Explanation

The `\s` pattern matches whitespace characters such as spaces, tabs, and newlines.

## 6️⃣ `\S` — Match Non-Whitespace

````python
text = "Python Programming"

print(re.findall(r"\S", text))
````

### Explanation

The `\S` pattern matches every character except whitespace.

## 7️⃣ `^` — Starts With

````python
text = "Python Programming"

print(re.search(r"^Python", text))
````

### Explanation

The `^` symbol checks whether the string starts with the specified pattern.

## 8️⃣ `$` — Ends With

````python
text = "Python Programming"

print(re.search(r"Programming$", text))
````

### Explanation

The `$` symbol checks whether the string ends with the specified pattern.

## 9️⃣ Character Set `[]`

````python
text = "Cat Bat Hat Rat"

print(re.findall(r"[CB]", text))
````

### Explanation

Square brackets `[]` match **any one character** inside the brackets.

In this example, only **C** and **B** are matched.

## 🌍 Real-World Example: Extract Phone Number

````python
text = "Contact: 9876543210"

print(re.findall(r"\d", text))
````

### Explanation

The `\d` pattern extracts all digits from the text.

## 🌍 Real-World Example: Check File Extension

````python

filename = "report.pdf"

print(re.search(r"\.pdf$", filename))
````

### Explanation

The regular expression checks whether the filename ends with the `.pdf` extension.

## 📋 Quick Reference Table

| Symbol | Meaning            | Example    |
| ------ | ------------------ | ---------- |
| `\d`   | Digit              | `5`        |
| `\D`   | Non-digit          | `A`        |
| `\w`   | Letter, Digit, `_` | `Python_1` |
| `\W`   | Special Character  | `@`        |
| `\s`   | Whitespace         | `" "`      |
| `\S`   | Non-whitespace     | `P`        |
| `^`    | Start of String    | `^Python`  |
| `$`    | End of String      | `Python$`  |
| `[]`   | Character Set      | `[ABC]`    |


## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting to import the `re` module before using Regular Expressions.
- ❌ Confusing `re.search()` with `re.match()`.
- ❌ Forgetting to use raw strings (`r"pattern"`) for regex patterns.
- ❌ Using `findall()` when only the first match is needed.
- ❌ Writing incorrect regex patterns due to missing escape characters.
- ❌ Assuming regex is always the best solution for simple string operations.

## 💡 Best Practices

Follow these best practices while using Regular Expressions:

- ✅ Always import the `re` module before using regex functions.
- ✅ Use raw strings (`r"..."`) to write regex patterns.
- ✅ Keep regex patterns simple and readable.
- ✅ Test regex patterns with sample data before using them in applications.
- ✅ Use descriptive variable names for patterns.
- ✅ Use normal string methods when regex is not required.
