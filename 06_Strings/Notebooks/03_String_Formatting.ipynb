# 🎨 String Formatting in Python

## 📖 Introduction

String formatting is the process of inserting values into a string in a clean and readable way.

Instead of combining strings manually using the `+` operator, Python provides several formatting techniques that make code easier to write and understand.

The three most commonly used formatting methods are:

- `%` Formatting (Old Style)
- `str.format()` Method
- `f-Strings` (Formatted String Literals)

## 🎯 Learning Objectives

By the end of this notebook, you will be able to:

- Understand different string formatting techniques.
- Use `%` formatting.
- Use the `str.format()` method.
- Use f-Strings.
- Format numbers with precision.
- Align text for better output.

# 🏷️ 1. `%` Formatting (Old Style)

`%` formatting is the oldest way of formatting strings in Python.

Although it is still supported, modern Python programs generally prefer **f-Strings** because they are more readable and easier to write.

## 📝 Syntax

"format_string" % values

- **format_string** → The string containing formatting placeholders.
- **%** → Formatting operator.
- **values** → The values inserted into the placeholders.

## 💻 Example 1: Formatting a String

````python
name = "Nikita"

print("Hello, %s!" % name)
````

### Explanation

The `%s` placeholder is replaced with the value of the variable `name`.

## 💻 Example 2: Formatting an Integer

````python
age = 21

print("Age: %d" % age)
````

### Explanation

The `%d` placeholder is used to display integer values.

## 💻 Example 3: Formatting a Float

````python
pi = 3.14159265

print("Value: %.2f" % pi)
````

### Explanation

The `%.2f` placeholder displays a floating-point number with **2 decimal places**.

## 🌍 Real-World Example: Student Report

````python
name = "Nikita"
marks = 92

print("Student: %s | Marks: %d" % (name, marks))
````

### Explanation

Multiple values can be inserted into a string by providing them as a tuple.

# 📝 2. str.format() Method

The `str.format()` method is more flexible than `%` formatting.

It allows values to be inserted into placeholders using `{}`.

## 📝 Syntax

"{}".format(value)

- `{}` → Placeholder for the value.
- `format()` → Inserts the specified value into the placeholder.

## 💻 Example 1: Single Value

````python
name = "Nikita"

print("Hello, {}!".format(name))
````

### Explanation

The `{}` placeholder is replaced with the value passed to the `format()` method.

## 💻 Example 2: Multiple Values

````python
name = "Nikita"
age = 21

print("Name: {} | Age: {}".format(name, age))
````

### Explanation

Multiple placeholders are filled in the same order as the values passed to `format()`.

## 💻 Example 3: Named Placeholders

````python
print("Name: {name}, Age: {age}".format(name="Nikita", age=21))
````

### Explanation

Named placeholders improve readability, especially when formatting many values.

# 🚀 3. f-Strings (Formatted String Literals)

f-Strings were introduced in **Python 3.6**.

They provide the most readable and efficient way to insert variables and expressions into strings.

An `f` or `F` is placed before the opening quotation mark, and variables or expressions are enclosed inside curly braces `{}`.

## 📝 Syntax

f"Text {variable}"

- `f` → Indicates that the string is an f-String.
- `{}` → Placeholder used to insert variables or expressions directly into the string.

## 💻 Example 1: Display a Variable

````python
name = "Nikita"

print(f"Hello, {name}!")
````

### Explanation

The variable `name` is inserted directly into the string using curly braces `{}`.

## 💻 Example 2: Multiple Variables

````python
name = "Nikita"
age = 21

print(f"Name: {name}, Age: {age}")
````

### Explanation

Multiple variables can be inserted into the same string, making the code concise and easy to read.

## 💻 Example 3: Expressions

````python
a = 15
b = 25

print(f"Sum = {a + b}")
````

### Explanation

f-Strings can evaluate expressions directly inside the curly braces.

## 💻 Example 4: Method Calls

````python
name = "nikita"

print(f"Formatted Name: {name.title()}")
````

### Explanation

Methods can be called directly inside an f-String, making formatting more convenient.

## 💻 Example 5: Float Formatting

````python
pi = 3.1415926535

print(f"Value of Pi: {pi:.2f}")
````

### Explanation

The format specifier `.2f` displays the floating-point number with **2 decimal places**.

## 🌍 Real-World Example: Student Report Card

````python
name = "Nikita"
marks = 91.75

print(f"Student: {name} | Marks: {marks:.2f}")
````

### Explanation

f-Strings make it easy to combine text, variables, and formatted numbers into a single readable output.

# 📏 String Alignment

Alignment is useful when displaying tables, reports, or formatted output.

Python provides three alignment options:

- `<` → Left Align
- `>` → Right Align
- `^` → Center Align

## 1️⃣ Left Alignment

````python
name = "Python"

print(f"|{name:<15}|")
````

### Explanation

The `<15` format specifier left-aligns the text within a field of **15 characters**.

## 2️⃣ Right Alignment

````python
name = "Python"

print(f"|{name:>15}|")
````

### Explanation

The `>15` format specifier right-aligns the text within a field of **15 characters**.

## 3️⃣ Center Alignment

````python
name = "Python"

print(f"|{name:^15}|")
````

### Explanation

The `^15` format specifier centers the text within a field of **15 characters**.

## 🌍 Real-World Example: Student Table

````python
name = "Nikita"
marks = 92

print(f"|{'Name':<15}|{'Marks':>8}|")
print("-" * 26)
print(f"|{name:<15}|{marks:>8}|")
````

### Explanation

Alignment specifiers make tabular output neat and easy to read, which is useful for reports and console applications.

# 🔢 Number Formatting

Python allows numbers to be displayed in different formats using format specifiers.

## 💻 Example 1: Two Decimal Places

````python
price = 2499.9876

print(f"Price: ₹{price:.2f}")
````

### Explanation

The `.2f` format specifier rounds the number and displays exactly **2 decimal places**.

## 💻 Example 2: Comma Separator

````python
population = 1425000000

print(f"Population: {population:,}")
````

### Explanation

The `:,` format specifier adds commas to large numbers, improving readability.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using the `+` operator instead of string formatting for complex output.
- ❌ Forgetting to add `f` before an f-String.
- ❌ Using the wrong placeholder in `%` formatting (such as `%d` for a string).
- ❌ Mismatching the number of placeholders and values in `str.format()`.
- ❌ Forgetting to close curly braces `{}` in f-Strings.
- ❌ Using formatting methods that reduce code readability.

## 💡 Best Practices

Follow these best practices while formatting strings:

- ✅ Prefer **f-Strings** in modern Python programs.
- ✅ Use meaningful variable names.
- ✅ Format floating-point numbers using `.2f` or an appropriate precision.
- ✅ Use alignment when displaying tables or reports.
- ✅ Keep formatted strings simple and readable.
