# 🔤 String Basics in Python

## 📖 Introduction

A **string** is a sequence of characters used to store and manipulate text.

Strings are one of the most commonly used data types in Python. They are used to store names, sentences, passwords, email addresses, file paths, and many other types of textual data.

In Python, strings are **immutable**, which means their contents cannot be changed after they are created.

## 🤔 What is a String?

A string is a collection of characters enclosed within quotation marks.

Python allows strings to be written using:

- Single quotes (`' '`)
- Double quotes (`" "`)
- Triple single quotes (`''' '''`)
- Triple double quotes (`""" """`)

## 💻 Example 1: Single Quotes

````python
language = 'Python'

print(language)
````

### Explanation

The string is enclosed using **single quotation marks**.

Single quotes are commonly used for short strings.

## 💻 Example 2: Double Quotes

````python
message = "Hello, World!"

print(message)
````

### Explanation

The string is enclosed using **double quotation marks**.

Single and double quotes work the same way in Python.

## 💻 Example 3: Triple Quotes

````python
paragraph = """
Python is easy to learn.

It is widely used in Data Science.
"""

print(paragraph)
````

### Explanation

Triple quotes are mainly used for:

- Multi-line strings
- Long text
- Documentation (Docstrings)

## 💻 Example 4: Quotes Inside Strings

````python
message1 = "I'm learning Python."

message2 = 'He said, "Python is amazing!"'

print(message1)
print(message2)
````

### Explanation

Using different quotation marks allows quotation marks to appear inside a string without causing syntax errors.

## 💻 Example 5: Empty String

````python
text = ""

print(text)
print(type(text))
````

### Explanation

An empty string contains no characters but is still a valid string object.

## 💻 Example 6: String Length

````python
language = "Python"

print(len(language))
````

### Explanation

The `len()` function returns the total number of characters in the string, including spaces and special characters.

# 🔢 String Indexing

## 🤔 What is String Indexing?

**Indexing** is used to access individual characters in a string.

Each character has a unique position called an **index**.

Python supports:

- **Positive Indexing** (Left → Right)
- **Negative Indexing** (Right → Left)

### Example String

```text
P  y  t  h  o  n
0  1  2  3  4  5

P  y  t  h  o  n
-6 -5 -4 -3 -2 -1
```

## ➡️ Positive Indexing

````python
language = "Python"

print(language[0])
print(language[1])
print(language[2])
````

### Explanation

Positive indexing starts from **0**.

The first character has index `0`, the second has index `1`, and so on.

## ⬅️ Negative Indexing

````python
language = "Python"

print(language[-1])
print(language[-2])
print(language[-3])
````

### Explanation

Negative indexing starts from the **end** of the string.

The last character has index `-1`, the second last has index `-2`, and so on.

## 💻 Example: Accessing Different Characters

````python
text = "Programming"

print(text[0])
print(text[5])
print(text[-1])
print(text[-4])
````

### Explanation

You can access any character in a string using either positive or negative indexing.

## ⚠️ IndexError

````python
language = "Python"

print(language[10])
````

### Explanation

If you try to access an index that doesn't exist, Python raises an **IndexError**.

# ✂️ String Slicing

## 🤔 What is String Slicing?

**Slicing** is used to extract a portion of a string.

Instead of accessing one character, slicing returns a **new string** containing multiple characters.

## 📝 Syntax

string[start : stop : step]

- **start** → Starting index (included).
- **stop** → Ending index (excluded).
- **step** → Number of positions to move each time.

## 💻 Example 1: Basic Slicing

````python
language = "Python"

print(language[0:4])
````

### Explanation

The slice starts from index **0** and stops before index **4**.

Characters at positions `0`, `1`, `2`, and `3` are included.

## 💻 Example 2: Omitting Start Index

````python
language = "Python"

print(language[:4])
````

### Explanation

If the **start** index is omitted, Python automatically starts from index **0**.

## 💻 Example 3: Omitting Stop Index

````python
language = "Python"

print(language[2:])
````

### Explanation

If the **stop** index is omitted, Python continues slicing until the end of the string.

## 💻 Example 4: Complete Copy

````python
language = "Python"

print(language[:])
````

### Explanation

Using `[:]` creates a copy of the entire string.

## 💻 Example 5: Using Step

````python
language = "Python"

print(language[::2])
````

### Explanation

The **step** value of `2` means every second character is selected.

## 💻 Example 6: Reverse a String

````python
language = "Python"

print(language[::-1])
````

### Explanation

A step value of `-1` moves from right to left, creating a reversed string.

## 🌍 Real-World Example: Email Username


````python
email = "nikita@gmail.com"

username = email[:email.index("@")]

print(username)
````

### Explanation

The program extracts the username by taking all characters before the `@` symbol using string slicing.

# 🔒 String Immutability

## 🤔 What is String Immutability?

Strings in Python are **immutable**, which means **their contents cannot be changed after they are created**.

If you perform an operation that appears to modify a string, Python actually creates a **new string** instead of changing the original one.

## 💻 Example 1: Attempting to Modify a Character

````python
language = "Python"

language[0] = "J"
````

### Explanation

A string does not allow individual characters to be changed.

Trying to modify a character using its index raises a **TypeError**.

## 💻 Example 2: Creating a New String

````python
language = "Python"

new_language = "J" + language[1:]

print(language)
print(new_language)
````

### Explanation

The original string remains unchanged.

A new string named `new_language` is created with the updated value.

## 💻 Example 3: Using `replace()`

````python
language = "Python"

updated = language.replace("Python", "Java")

print(language)
print(updated)
````

### Explanation

The `replace()` method does not modify the original string.

Instead, it returns a **new string** with the replacement.

## 💻 Example 4: String Concatenation

````python
text = "Hello"

text = text + " World"

print(text)
````

### Explanation

The original string is not modified.

Python creates a new string `"Hello World"` and assigns it back to the variable `text`.

## 🌍 Real-World Example: Username Formatting

````python
username = "nikita"

formatted_username = username.capitalize()

print(username)
print(formatted_username)
````

### Explanation

The original username remains unchanged.

The `capitalize()` method returns a new formatted string.

## 🌍 Real-World Example: File Extension

````python
filename = "report.txt"

new_filename = filename.replace(".txt", ".pdf")

print(filename)
print(new_filename)
````

### Explanation

The original filename is preserved.

A new string with the updated extension is created.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Trying to modify a character using indexing.
- ❌ Confusing indexing with slicing.
- ❌ Forgetting that the stop index in slicing is excluded.
- ❌ Assuming string methods modify the original string.
- ❌ Accessing an index that does not exist.

## 💡 Best Practices

Follow these best practices while working with strings:

- ✅ Use meaningful variable names.
- ✅ Check string length before accessing indexes.
- ✅ Use slicing instead of unnecessary loops.
- ✅ Remember that strings are immutable.
- ✅ Store the result of string methods in a new variable when needed.
