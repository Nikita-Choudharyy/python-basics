# ⚖️ Comparison and Logical Operators in Python

## 📖 Introduction

Comparison operators are used to compare two values and return a Boolean result (`True` or `False`).

Logical operators are used to combine multiple conditions and also return a Boolean result.

Together, these operators play a vital role in decision-making and are commonly used with conditional statements and loops.

- Apply these operators in real-world scenarios.

## ⚖️ Comparison Operators

Comparison operators compare two values and return either `True` or `False`.

| Operator | Description | Example |
|----------|-------------|---------|
| `==` | Equal to | `10 == 10` |
| `!=` | Not equal to | `10 != 5` |
| `>` | Greater than | `10 > 5` |
| `<` | Less than | `5 < 10` |
| `>=` | Greater than or equal to | `10 >= 10` |
| `<=` | Less than or equal to | `5 <= 10` |

### Example: Comparison Operators

```python
age = 20

print(age >= 18)
print(age < 18)
print(age == 20)
print(age != 25)
```

### Explanation

Each comparison checks the relationship between two values and returns either `True` or `False`.

## 🧠 Logical Operators

Logical operators are used to combine or modify conditions.

Python provides three logical operators:

| Operator | Description |
|----------|-------------|
| `and` | Returns `True` if both conditions are `True`. |
| `or` | Returns `True` if at least one condition is `True`. |
| `not` | Reverses the Boolean value. |

### `and` Operator

```python
age = 22
has_license = True

print(age >= 18 and has_license)
```

The `and` operator returns `True` only when **both conditions** are `True`.

### `or` Operator

```python
is_student = False
has_discount_coupon = True

print(is_student or has_discount_coupon)
```

The `or` operator returns `True` if **at least one condition** is `True`.

### `not` Operator

```python
is_logged_in = False

print(not is_logged_in)
```

The `not` operator reverses the Boolean value.

- `not True` → `False`
- `not False` → `True`

## 🌍 Real-World Example

```python
age = 20
has_id = True

if age >= 18 and has_id:
    print("Entry Allowed")
else:
    print("Entry Denied")
```

### Explanation

The person is allowed entry because:

- The age is **18 or above**.
- The person has a valid ID.

Since both conditions are `True`, the `and` operator returns `True`.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using `=` instead of `==` for comparison.
- ❌ Confusing `and` with `or`.
- ❌ Forgetting to use parentheses in complex conditions.
- ❌ Assuming `not` changes the variable instead of the Boolean result.

## 💡 Best Practices

Follow these best practices while using comparison and logical operators:

- ✅ Use meaningful variable names.
- ✅ Keep conditions simple and readable.
- ✅ Use parentheses when combining multiple conditions.
- ✅ Test different input values to verify your logic.
