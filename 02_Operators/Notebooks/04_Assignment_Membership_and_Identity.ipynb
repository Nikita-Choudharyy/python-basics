# 📝 Assignment, Membership, and Identity Operators in Python

## 📖 Introduction

Assignment, Membership, and Identity operators are commonly used in Python programs.

- **Assignment Operators** are used to assign and update values.
- **Membership Operators** check whether a value exists in a sequence.
- **Identity Operators** check whether two variables refer to the same object in memory.

Although these operators serve different purposes, they are frequently used while writing Python programs.

- Use assignment operators.
- Check membership using `in` and `not in`.
- Compare object identity using `is` and `is not`.
- Understand practical use cases of these operators.

# 📝 Assignment Operators

Assignment operators are used to assign values to variables and update existing values.

| Operator | Example | Equivalent To |
|----------|---------|---------------|
| `=` | `x = 10` | Assign value |
| `+=` | `x += 5` | `x = x + 5` |
| `-=` | `x -= 5` | `x = x - 5` |
| `*=` | `x *= 5` | `x = x * 5` |
| `/=` | `x /= 5` | `x = x / 5` |
| `//=` | `x //= 5` | `x = x // 5` |
| `%=` | `x %= 5` | `x = x % 5` |
| `**=` | `x **= 2` | `x = x ** 2` |

### Example: Assignment Operators

```python
salary = 50000

salary += 5000

print(salary)

salary -= 2000

print(salary)
```

### Explanation

The `+=` operator increases the value of a variable, while the `-=` operator decreases it.

These operators provide a shorter and more readable way to update variable values.

# 📦 Membership Operators

Membership operators are used to check whether a value exists in a sequence such as a string, list, tuple, or dictionary.

Python provides two membership operators:

| Operator | Description |
|----------|-------------|
| `in` | Returns `True` if the value exists. |
| `not in` | Returns `True` if the value does not exist. |

### Example: Membership Operators

```python
fruits = ["Apple", "Banana", "Mango"]

print("Apple" in fruits)

print("Orange" not in fruits)
```

### Explanation

- `"Apple" in fruits` checks whether `"Apple"` exists in the list.
- `"Orange" not in fruits` checks whether `"Orange"` is absent from the list.

# 🆔 Identity Operators

Identity operators are used to check whether two variables refer to the **same object in memory**.

Python provides two identity operators:

| Operator | Description |
|----------|-------------|
| `is` | Returns `True` if both variables refer to the same object. |
| `is not` | Returns `True` if both variables refer to different objects. |

### Example: Identity Operators

```python
list1 = [10, 20, 30]

list2 = list1

print(list1 is list2)

print(list1 is not list2)
```

### Explanation

Both `list1` and `list2` refer to the same object in memory.

Therefore:

- `list1 is list2` returns `True`.
- `list1 is not list2` returns `False`.

## 🌍 Real-World Example

```python
username = "Nikita"

registered_users = ["Aman", "Nikita", "Riya"]

print(username in registered_users)
```

### Explanation

The `in` operator checks whether the username exists in the list of registered users.

Since `"Nikita"` is present, the result is `True`.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using `==` instead of `is`.
- ❌ Using `is` to compare numbers or strings instead of `==`.
- ❌ Confusing `in` with `is`.
- ❌ Forgetting that assignment operators update the existing value of a variable.

## 💡 Best Practices

Follow these best practices:

- ✅ Use `==` to compare values.
- ✅ Use `is` only when checking object identity.
- ✅ Use membership operators with sequences like lists, tuples, strings, and dictionaries.
- ✅ Use assignment operators to write shorter and more readable code.
