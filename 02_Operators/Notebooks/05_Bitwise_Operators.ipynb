# 💻 Bitwise Operators in Python

## 📖 Introduction

Bitwise operators perform operations directly on the **binary (bit)** representation of integers.

Unlike arithmetic operators, which work with decimal numbers, bitwise operators compare or manipulate individual bits.

They are commonly used in low-level programming, networking, embedded systems, encryption, and performance optimization.

## 🎯 Learning Objectives

By the end of this notebook, you will be able to:

- Understand what bitwise operators are.
- Learn how decimal numbers are represented in binary.
- Perform bitwise operations in Python.
- Understand the purpose of each bitwise operator.

## 🧠 Before Learning Bitwise Operators

Bitwise operators work on **binary numbers** (0s and 1s).

For example:

| Decimal | Binary |
|---------:|:------:|
| 5 | `0101` |
| 3 | `0011` |
| 8 | `1000` |

Python automatically converts decimal numbers into binary before performing bitwise operations.

## 📋 Bitwise Operators

| Operator | Name | Example |
|----------|------|---------|
| `&` | Bitwise AND | `5 & 3` |
| `|` | Bitwise OR | `5 | 3` |
| `^` | Bitwise XOR | `5 ^ 3` |
| `~` | Bitwise NOT | `~5` |
| `<<` | Left Shift | `5 << 1` |
| `>>` | Right Shift | `5 >> 1` |

## 🔹 Bitwise AND (`&`)

The **AND (`&`)** operator returns `1` only when **both corresponding bits are `1`**.

Example:

```text
  5  → 0101
  3  → 0011
---------------
&     0001
```

The binary result `0001` is equal to **1** in decimal.

````python
print(5 & 3)
````

### Explanation

Python converts both numbers to binary, performs the **AND** operation bit by bit, and converts the result back to decimal.

## 🔹 Bitwise OR (`|`)

The **OR (`|`)** operator returns `1` if **at least one corresponding bit is `1`**.

Example:

```text
  5  → 0101
  3  → 0011
---------------
|     0111
```

The binary result `0111` is equal to **7** in decimal.

````python
print(5 | 3)
````

### Explanation

The OR operator compares each pair of bits and returns `1` whenever at least one of the bits is `1`.

## 🔹 Bitwise XOR (`^`)

The **XOR (`^`)** operator returns `1` only when the corresponding bits are **different**.

Example:

```text
  5  → 0101
  3  → 0011
---------------
^     0110
```

The binary result `0110` is equal to **6** in decimal.

````python
print(5 ^ 3)
````

### Explanation

The XOR operator compares each pair of bits.

- Same bits (`0-0` or `1-1`) → `0`
- Different bits (`0-1` or `1-0`) → `1`

## 🔹 Bitwise NOT (`~`)

The **NOT (`~`)** operator flips every bit.

Unlike other bitwise operators, it works with **only one operand**.

For positive numbers, Python returns a negative value because it uses **two's complement representation** for integers.

````python
print(~5)
````

### Explanation

For beginners, remember this simple rule:

```text
~n = -(n + 1)
```

Example:

```text
~5 = -(5 + 1)
   = -6
```

You don't need to understand two's complement in depth right now. Just remember the formula above.

## 🔹 Left Shift (`<<`)

The left shift operator moves all bits to the left by the specified number of positions.

Each left shift by one position is equivalent to multiplying the number by **2**.

````python
print(5 << 1)

print(5 << 2)
````

### Explanation

```text
5 << 1  → 10

5 << 2  → 20
```

Every left shift doubles the value.

## 🔹 Right Shift (`>>`)

The right shift operator moves all bits to the right by the specified number of positions.

Each right shift by one position is approximately equivalent to dividing the number by **2**.

````python
print(20 >> 1)

print(20 >> 2)
````

### Explanation

```text
20 >> 1 → 10

20 >> 2 → 5
```

Every right shift halves the value (for positive integers).

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using `^` for exponentiation (use `**` instead).
- ❌ Confusing `|` with the logical operator `or`.
- ❌ Confusing `&` with the logical operator `and`.
- ❌ Expecting `~5` to return `-5` instead of `-6`.

## 💡 Best Practices

Follow these best practices:

- ✅ Use bitwise operators only when needed.
- ✅ Understand the binary representation before applying bitwise operations.
- ✅ Use parentheses in complex expressions.
- ✅ Use `**` for exponentiation instead of `^`.
