# ➕ Arithmetic Operators in Python

## 📖 Introduction

Arithmetic operators are used to perform mathematical calculations on numbers.

They are among the most frequently used operators in Python and are essential for tasks such as calculations, data analysis, scientific computing, and algorithm development.

## 📋 Arithmetic Operators

| Operator | Description | Example |
|----------|-------------|---------|
| `+` | Addition | `10 + 5` |
| `-` | Subtraction | `10 - 5` |
| `*` | Multiplication | `10 * 5` |
| `/` | Division | `10 / 5` |
| `//` | Floor Division | `10 // 3` |
| `%` | Modulus | `10 % 3` |
| `**` | Exponent | `2 ** 3` |

## ➕ Addition (`+`)

The `+` operator adds two numbers and returns their sum.

```python
num1 = 15
num2 = 10

print(num1 + num2)
```

## ➖ Subtraction (`-`)

The `-` operator subtracts one value from another.

```python
balance = 5000
expense = 1750

print(balance - expense)
```

## ✖️ Multiplication (`*`)

The `*` operator multiplies two values.

```python
price = 250
quantity = 4

print(price * quantity)
```

## ➗ Division (`/`)

The `/` operator always returns a floating-point (`float`) result, even when the division is exact.

```python
total_marks = 450
subjects = 5

print(total_marks / subjects)
```

## ➗ Floor Division (`//`)

The floor division operator (`//`) divides two numbers and returns the **quotient without the decimal part**.

It rounds the result **down** to the nearest whole number.

```python
num1 = 17
num2 = 5

print(num1 // num2)
```

### Explanation

When `17` is divided by `5`, the result is `3.4`.

The floor division operator returns only the whole number part by rounding down, which is `3`.

## 🧮 Modulus (`%`)

The modulus operator (`%`) returns the **remainder** after division.

It is commonly used to check whether a number is even or odd.

```python
num = 17

print(num % 5)
```

### Real-World Example: Even or Odd

```python
number = 12

print(number % 2)
```

If the remainder is:

- `0` → The number is **Even**.
- `1` → The number is **Odd**.

## 🔼 Exponent (`**`)

The exponent operator (`**`) raises a number to the power of another number.

```python
print(2 ** 3)

print(5 ** 2)
```

### Explanation

- `2 ** 3` means **2 raised to the power of 3**, which equals `8`.
- `5 ** 2` means **5 squared**, which equals `25`.

## 🌍 Real-World Example

```python
product_price = 750
quantity = 3

total_bill = product_price * quantity

print("Total Bill: ₹", total_bill)
```

This example calculates the total bill by multiplying the price of one product by the quantity purchased.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Confusing `/` with `//`.
- ❌ Forgetting that `/` always returns a `float`.
- ❌ Using `%` without understanding that it returns the remainder.
- ❌ Writing `^` instead of `**` for exponentiation.

## 💡 Best Practices

Follow these best practices while using arithmetic operators:

- ✅ Use meaningful variable names.
- ✅ Choose the correct operator for the calculation.
- ✅ Use parentheses to make complex expressions easier to read.
- ✅ Test your calculations with different values.
