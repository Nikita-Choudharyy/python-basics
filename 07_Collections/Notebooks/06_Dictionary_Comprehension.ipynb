# 📖 Dictionary Comprehension in Python

## 📖 Introduction

Dictionary Comprehension is a concise and elegant way to create dictionaries in Python.

Instead of writing multiple lines of code using loops and assignments, Dictionary Comprehension allows you to create a complete dictionary in a single line.

It is widely used in Data Science, Machine Learning, Data Analysis, Web Development, and Automation for transforming and organizing data efficiently.

## 🤔 What is Dictionary Comprehension?

Dictionary Comprehension is a compact way to create dictionaries by generating **key-value pairs** from an iterable.

It works similarly to List Comprehension but produces a dictionary instead of a list.

### General Syntax

```python
{key_expression: value_expression for item in iterable}
```

With a condition:

```python
{key_expression: value_expression for item in iterable if condition}
```

## ✨ Why Use Dictionary Comprehension?

Dictionary Comprehension provides several benefits:

- Produces cleaner and shorter code.
- Improves readability.
- Reduces the need for loops.
- Efficiently transforms data.
- Frequently used in Machine Learning preprocessing and data manipulation.

## 💻 Example 1: Create a Dictionary

````python
numbers = {x: x for x in range(1, 6)}

print(numbers)
````

### Explanation

Each number becomes both the key and the value in the dictionary.

## 💻 Example 2: Squares of Numbers

````python
squares = {x: x**2 for x in range(1, 6)}

print(squares)
````

### Explanation

Each number is used as the key, while its square becomes the corresponding value.

## 💻 Example 3: Cubes of Numbers

````python
cubes = {x: x**3 for x in range(1, 6)}

print(cubes)
````

### Explanation

The cube of each number is stored as the dictionary value.

## 💻 Example 4: Length of Words

````python
words = ["Python", "Machine", "Learning", "AI"]

word_lengths = {word: len(word) for word in words}

print(word_lengths)
````

### Explanation

Each word becomes a key, while its length becomes the corresponding value.

## 💻 Example 5: Convert to Uppercase

````python
names = ["nikita", "rahul", "aman"]

upper_names = {name: name.upper() for name in names}

print(upper_names)
````

### Explanation

Each original name becomes the key, while its uppercase version becomes the value.

## 💻 Example 6: Dictionary from Two Lists

````python
subjects = ["Python", "SQL", "Machine Learning"]
marks = [95, 90, 98]

student_marks = {subject: mark for subject, mark in zip(subjects, marks)}

print(student_marks)
````

### Explanation

The `zip()` function combines two lists, and Dictionary Comprehension converts them into key-value pairs.

## 🌍 Real-World Example: Student Grades

````python
marks = {
    "Nikita": 92,
    "Rahul": 75,
    "Aman": 84
}

grades = {
    name: "Pass" if mark >= 40 else "Fail"
    for name, mark in marks.items()
}

print(grades)
````

### Explanation

Schools often need to generate student results automatically.

Dictionary Comprehension can transform marks into pass/fail status quickly and efficiently.

# 🔍 Conditional Dictionary Comprehension

Sometimes, we want to include only specific key-value pairs in a dictionary.

Dictionary Comprehension allows us to filter data using an **`if` condition**.

### Syntax

```python
{key: value for item in iterable if condition}
```

## 💻 Example 1: Even Numbers and Their Squares

````python
even_squares = {x: x**2 for x in range(1, 11) if x % 2 == 0}

print(even_squares)
````

### Explanation

Only even numbers satisfy the condition and become keys in the dictionary.

Their squares are stored as the corresponding values.

## 💻 Example 2: Words Longer Than 5 Characters

````python
words = ["Python", "AI", "Machine", "SQL", "Learning"]

word_lengths = {
    word: len(word)
    for word in words
    if len(word) > 5
}

print(word_lengths)
````

### Explanation

Only words with more than **5 characters** are included in the dictionary.

# 🔀 Using if-else in Dictionary Comprehension

The **`if-else` expression** allows us to transform dictionary values based on a condition.

### Syntax

```python
{
    key: value_if_true if condition else value_if_false
    for item in iterable
}
```

## 💻 Example 1: Student Result

````python
marks = {
    "Nikita": 92,
    "Rahul": 35,
    "Aman": 75,
    "Priya": 28
}

results = {
    name: "Pass" if mark >= 40 else "Fail"
    for name, mark in marks.items()
}

print(results)
````

### Explanation

Every student's marks are evaluated.

- Marks **40 or above** → `"Pass"`
- Below **40** → `"Fail"`

## 💻 Example 2: Discount Eligibility

````python
customers = {
    "Amit": 1200,
    "Riya": 650,
    "Rahul": 2000,
    "Neha": 500
}

discount = {
    name: "Eligible" if amount >= 1000 else "Not Eligible"
    for name, amount in customers.items()
}

print(discount)
````

### Explanation

An online shopping application can automatically determine which customers qualify for a discount based on their purchase amount.

# 🛠️ Using Functions in Dictionary Comprehension

Functions can be applied to keys or values while creating a dictionary.

This makes data transformation simple and efficient.

## 💻 Example 1: Convert Names to Uppercase

````python
names = ["nikita", "rahul", "aman"]

result = {
    name: name.upper()
    for name in names
}

print(result)
````

### Explanation

The `upper()` method is applied to every value before it is stored in the dictionary.

## 💻 Example 2: Absolute Values

````python
numbers = [-5, -2, 4, 8]

absolute_numbers = {
    number: abs(number)
    for number in numbers
}

print(absolute_numbers)
````

### Explanation

The built-in `abs()` function converts every number into its absolute value.

# 🧩 Nested Dictionary Comprehension

Nested Dictionary Comprehension can generate complex dictionaries using nested loops.

Although powerful, it should be used carefully to maintain code readability.

## 💻 Example 1: Multiplication Table

````python
table = {
    number: {
        multiplier: number * multiplier
        for multiplier in range(1, 6)
    }
    for number in range(1, 4)
}

print(table)
````

### Explanation

The outer comprehension creates the main dictionary, while the inner comprehension creates another dictionary containing multiplication tables.

## 🌍 Real-World Example: Product Price with GST

````python
prices = {
    "Laptop": 50000,
    "Mouse": 800,
    "Keyboard": 1500
}

final_prices = {
    product: price * 1.18
    for product, price in prices.items()
}

print(final_prices)
````

### Explanation

Businesses often need to calculate prices including GST.

Dictionary Comprehension updates every product price in a single line, making the code concise and easy to maintain.

## 📊 Quick Reference Table

| Syntax                                                                  | Purpose                                |
| ----------------------------------------------------------------------- | -------------------------------------- |
| `{k: v for x in iterable}`                                              | Create a dictionary                    |
| `{k: v for x in iterable if condition}`                                 | Filter key-value pairs                 |
| `{k: value_if_true if condition else value_if_false for x in iterable}` | Transform values using `if-else`       |
| `{word: len(word) for word in words}`                                   | Apply a function to values             |
| `{k: v for k, v in zip(keys, values)}`                                  | Create a dictionary from two iterables |


## 📊 List Comprehension vs Dictionary Comprehension

| Feature           | List Comprehension                  | Dictionary Comprehension               |
| ----------------- | ----------------------------------- | -------------------------------------- |
| Output            | List                                | Dictionary                             |
| Syntax            | `[expression for item in iterable]` | `{key: value for item in iterable}`    |
| Stores            | Values                              | Key-value pairs                        |
| Filtering         | ✅ Yes                               | ✅ Yes                                  |
| `if-else` Support | ✅ Yes                               | ✅ Yes                                  |
| Best Use Case     | Creating and transforming lists     | Creating and transforming dictionaries |


## ⚠️ Common Mistakes

Beginners often make these mistakes while using Dictionary Comprehension:

- ❌ Forgetting the colon (`:`) between the key and value expressions.
- ❌ Placing the `if` condition before the `for` loop.
- ❌ Confusing filtering (`if`) with value transformation (`if-else`).
- ❌ Creating very complex nested comprehensions that reduce readability.
- ❌ Assuming Dictionary Comprehension preserves duplicate keys (duplicate keys are overwritten).
- ❌ Using mutable objects such as lists as dictionary keys.

## 💡 Best Practices

Follow these best practices while using Dictionary Comprehension:

- ✅ Keep dictionary comprehensions short and readable.
- ✅ Use meaningful variable names.
- ✅ Use conditions only when necessary.
- ✅ Prefer a normal `for` loop if the logic becomes too complex.
- ✅ Apply built-in functions for simple data transformations.
- ✅ Test the output with a small dataset before using it on large datasets.
