# 📋 Lists in Python

## 📖 Introduction

A **list** is one of the most commonly used collection data types in Python.

Lists are used to store multiple values in a single variable.

Unlike strings, lists are **mutable**, which means their elements can be added, removed, or modified after the list is created.

Lists can store different types of data, including integers, floating-point numbers, strings, Boolean values, and even other lists.

## 🤔 What is a List?

A list is an **ordered, mutable collection** that allows duplicate values.

Lists are created using square brackets `[]`.

They are one of the most flexible data structures in Python because they can store multiple values of different data types in a single collection.

## ✨ Features of Lists

Python lists have the following features:

- Ordered collection
- Mutable (can be modified)
- Allows duplicate values
- Stores multiple data types
- Supports indexing and slicing
- Can contain nested lists
- Dynamic size (elements can be added or removed)

# 🔢 List Indexing

Lists are **ordered collections**, which means every element has a unique position called an **index**.

Python supports two types of indexing:

- **Positive Indexing**
- **Negative Indexing**

## 📍 Positive Indexing

Positive indexing starts from **0**.

Example:

```text
Index:      0        1        2        3
List :   ["Apple","Banana","Mango","Orange"]
```

````python
fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[3])
````

### Explanation

Each element is accessed using its index.

The first element always has index **0**.

## 📍 Negative Indexing

Negative indexing starts from the **end** of the list.

```text
Index:     -4      -3      -2      -1
List :   ["Apple","Banana","Mango","Orange"]
```

````python
fruits = ["Apple", "Banana", "Mango", "Orange"]

print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
print(fruits[-4])
````

### Explanation

Negative indexing is useful when you want to access elements from the end of the list.

## 💻 Example: Access Student Details

````python
student = ["Nikita", 20, "CSE", 92.5]

print("Name :", student[0])
print("Age :", student[1])
print("Branch :", student[2])
print("Marks :", student[3])
````

### Explanation

Lists can store related information together, making them useful for representing records such as student details.

## ⚠️ IndexError

````python
fruits = ["Apple", "Banana", "Mango"]

print(fruits[5])
````

### Explanation

Trying to access an index that does not exist raises an **IndexError**.

# ✂️ List Slicing

Slicing allows you to extract a portion of a list.

### Syntax

```python
list[start : stop : step]
```

- **start** → Starting index (inclusive)
- **stop** → Ending index (exclusive)
- **step** → Number of positions to skip

## 💻 Example 1: Basic Slicing

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[1:4])
````

### Explanation

The slice starts from index **1** and stops before index **4**.

## 💻 Example 2: From Beginning

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[:3])
````

### Explanation

If the starting index is omitted, Python starts from index **0**.

## 💻 Example 3: Till the End

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[2:])
````

### Explanation

If the ending index is omitted, Python slices until the last element.

## 💻 Example 4: Entire List

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[:])
````

### Explanation

Using `[:]` creates a shallow copy of the entire list.

## 💻 Example 5: Step Slicing

````python
numbers = [10, 20, 30, 40, 50, 60]

print(numbers[::2])
````

### Explanation

The step value **2** selects every second element from the list.

## 💻 Example 6: Reverse a List

````python
numbers = [10, 20, 30, 40, 50]

print(numbers[::-1])
````

### Explanation

A step value of **-1** reverses the list.

## 💻 Example 7: Accessing Nested Lists

````python
students = [
    ["Nikita", 92],
    ["Rahul", 88],
    ["Priya", 95]
]

print(students[0])
print(students[0][0])
print(students[1][1])
````

### Explanation

- `students[0]` accesses the first inner list.
- `students[0][0]` accesses the first element of the first inner list.
- `students[1][1]` accesses the marks of Rahul.

## 🌍 Real-World Example: Monthly Sales

````python
monthly_sales = [25000, 28000, 30000, 27000, 35000, 40000]

print("First Quarter:", monthly_sales[:3])
print("Last Three Months:", monthly_sales[-3:])
````

### Explanation

Businesses often use slicing to analyze sales for specific periods, such as quarterly performance or recent months.

# ➕ Adding Elements to a List

Python provides several methods to add new elements to a list.

The most commonly used methods are:

- `append()`
- `extend()`
- `insert()`

## 1️⃣ append()

The `append()` method adds a **single element** to the **end** of the list.

### Syntax

```python
list.append(element)
```

````python
fruits = ["Apple", "Banana"]

fruits.append("Mango")

print(fruits)
````

### Explanation

The `append()` method adds `"Mango"` as the last element of the list.

## 💻 Example: Append Different Data Types

````python
data = [10, 20]

data.append("Python")
data.append(True)

print(data)
````

### Explanation

Lists can store different data types, so `append()` can add integers, strings, Boolean values, or any other Python object.

## 2️⃣ extend()

The `extend()` method adds **multiple elements** from another iterable (such as a list, tuple, or set) to the end of the list.

### Syntax

```python
list.extend(iterable)
```

````python
fruits = ["Apple", "Banana"]

fruits.extend(["Mango", "Orange"])

print(fruits)
````

### Explanation

Unlike `append()`, `extend()` adds each element individually from the iterable.

## 💡 append() vs extend()

````python
numbers = [1, 2]

numbers.append([3, 4])
print(numbers)

numbers = [1, 2]

numbers.extend([3, 4])
print(numbers)
````

### Explanation

- `append()` adds the entire list as a single element.
- `extend()` adds each element separately.

## 3️⃣ insert()

The `insert()` method adds an element at a specific index.

### Syntax

```python
list.insert(index, element)
```

````python
fruits = ["Apple", "Banana", "Orange"]

fruits.insert(1, "Mango")

print(fruits)
````

### Explanation

The element is inserted at index **1**, and the remaining elements shift to the right.

# ❌ Removing Elements from a List

Python provides several methods to remove elements from a list.

Common methods include:

- `remove()`
- `pop()`
- `clear()`
- `del`

## 1️⃣ remove()

The `remove()` method removes the **first occurrence** of a specified value.

### Syntax

```python
list.remove(value)
```

````python
fruits = ["Apple", "Banana", "Mango"]

fruits.remove("Banana")

print(fruits)
````

### Explanation

The value `"Banana"` is removed from the list.

## 2️⃣ pop()

The `pop()` method removes and returns an element.

If no index is given, it removes the last element.

### Syntax

```python
list.pop(index)
```

````python
fruits = ["Apple", "Banana", "Mango"]

item = fruits.pop()

print(item)
print(fruits)
````

### Explanation

The last element is removed and stored in the variable `item`.

## 💻 Example: pop() with Index

````python
numbers = [10, 20, 30, 40]

print(numbers.pop(1))
print(numbers)
````

### Explanation

The element at index **1** is removed and returned.

## 3️⃣ clear()

The `clear()` method removes **all elements** from the list.

### Syntax

```python
list.clear()
```

````python
numbers = [10, 20, 30]

numbers.clear()

print(numbers)
````

### Explanation

The list still exists, but it becomes empty.

## 4️⃣ del Keyword

The `del` keyword can remove an element by index or delete the entire list.

### Syntax

```python
del list[index]
```

````python
numbers = [10, 20, 30, 40]

del numbers[2]

print(numbers)
````

### Explanation

The element at index **2** is permanently deleted from the list.

# ✏️ Updating List Elements

````python
fruits = ["Apple", "Banana", "Mango"]

fruits[1] = "Orange"

print(fruits)
````

### Explanation

Lists are mutable, so elements can be modified using their index.

## 🌍 Real-World Example: Student Attendance

````python
attendance = ["Nikita", "Rahul", "Priya"]

attendance.append("Aman")
attendance.remove("Rahul")

print(attendance)
````

### Explanation

A school's attendance system can add newly arrived students and remove absent or transferred students using list methods.

# 🔄 Common List Operations

Python provides several operators that can be used with lists.

Some common operations include:

- Concatenation (`+`)
- Repetition (`*`)
- Membership (`in`)
- Non-membership (`not in`)

## 1️⃣ Concatenation (`+`)

````python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = list1 + list2

print(result)
````

### Explanation

The `+` operator joins two lists and creates a new list.

## 2️⃣ Repetition (`*`)

````python
numbers = [10, 20]

print(numbers * 3)
````

### Explanation

The `*` operator repeats the list a specified number of times.

## 3️⃣ Membership Operator (`in`)

````python
fruits = ["Apple", "Banana", "Mango"]

print("Apple" in fruits)
print("Orange" in fruits)
````

### Explanation

The `in` operator checks whether an element exists in the list.

## 4️⃣ Non-Membership Operator (`not in`)

````python
fruits = ["Apple", "Banana", "Mango"]

print("Orange" not in fruits)
````

### Explanation

The `not in` operator checks whether an element is absent from the list.

# 📊 Useful Built-in Functions

## 1️⃣ len()

````python
numbers = [15, 25, 35, 45]

print(len(numbers))
````

### Explanation

`len()` returns the total number of elements in the list.

## 2️⃣ min()

````python
numbers = [15, 25, 35, 45]

print(min(numbers))
````

### Explanation

`min()` returns the smallest element from the list.

## 3️⃣ max()

````python
numbers = [15, 25, 35, 45]

print(max(numbers))
````

### Explanation

`max()` returns the largest element from the list.

## 4️⃣ sum()

````python
numbers = [10, 20, 30]

print(sum(numbers))
````

### Explanation

`sum()` returns the total sum of all numeric elements.

## 5️⃣ sorted()

````python
numbers = [40, 10, 30, 20]

print(sorted(numbers))
````

### Explanation

`sorted()` returns a new sorted list without changing the original list.

# 🔃 Useful List Methods

## 1️⃣ sort()

````python
numbers = [40, 10, 30, 20]

numbers.sort()

print(numbers)
````

### Explanation

`sort()` sorts the original list in ascending order.

## 💻 Descending Order

````python
numbers = [40, 10, 30, 20]

numbers.sort(reverse=True)

print(numbers)
````

### Explanation

Using `reverse=True` sorts the list in descending order.

## 2️⃣ reverse()

````python
numbers = [10, 20, 30, 40]

numbers.reverse()

print(numbers)
````

### Explanation

`reverse()` reverses the order of the existing list. It does **not** sort the list.

## 3️⃣ count()

````python
numbers = [10, 20, 10, 30, 10]

print(numbers.count(10))
````

### Explanation

`count()` returns how many times a value appears in the list.

## 4️⃣ index()

````python
fruits = ["Apple", "Banana", "Mango"]

print(fruits.index("Banana"))
````

### Explanation

`index()` returns the index of the **first occurrence** of the specified element.

## 5️⃣ copy()

````python
list1 = [10, 20, 30]

list2 = list1.copy()

print(list2)
````

### Explanation

`copy()` creates a shallow copy of the list, allowing you to work with a duplicate without modifying the original.

### Explanation

Schools and colleges can use built-in list functions to quickly analyze student performance, such as finding the highest score, lowest score, total marks, and sorted results.

## ⚠️ Common Mistakes

Beginners often make these mistakes while working with lists:

- ❌ Using an index that does not exist, resulting in an `IndexError`.
- ❌ Confusing `append()` and `extend()`.
- ❌ Using `remove()` for an element that is not in the list.
- ❌ Assuming `sort()` returns a new list (it modifies the original list).
- ❌ Forgetting that list indexing starts from `0`.
- ❌ Confusing `reverse()` with sorting in descending order.

## 💡 Best Practices

Follow these best practices while working with lists:

- ✅ Use meaningful variable names.
- ✅ Check whether an element exists before removing it.
- ✅ Use `append()` for adding a single element.
- ✅ Use `extend()` for adding multiple elements.
- ✅ Use `sorted()` when you want to keep the original list unchanged.
- ✅ Use list slicing instead of writing unnecessary loops when appropriate.
