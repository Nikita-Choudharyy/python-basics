# 📖 Dictionaries in Python

## 📖 Introduction

A **dictionary** is a built-in Python data structure used to store data as **key-value pairs**.

Each key in a dictionary is unique and is associated with a specific value.

Dictionaries are **mutable**, meaning you can add, update, or remove key-value pairs after the dictionary is created.

They are one of the most commonly used data structures in Python because they provide fast access to values using keys.

## 🤔 What is a Dictionary?

A **dictionary** stores information in the form of **key-value pairs**.

Each **key** uniquely identifies a value.

Unlike lists and tuples, dictionaries do not use numeric indexes for accessing data. Instead, values are accessed using their corresponding keys.

### General Syntax

```python
dictionary = {
    key1: value1,
    key2: value2,
    key3: value3
}
```

## ✨ Features of Dictionaries

Python dictionaries have the following features:

- Store data as key-value pairs
- Mutable (can be modified)
- Keys are unique
- Values can be duplicated
- Fast lookup using keys
- Can store different data types
- Can contain nested dictionaries

## 💻 Example 1: Creating a Dictionary

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student)
````

### Explanation

The dictionary contains three key-value pairs.

Each key is associated with one value.

## 💻 Example 2: Dictionary with Different Data Types

````python
data = {
    "id": 101,
    "name": "Nikita",
    "marks": 92.5,
    "passed": True
}

print(data)
````

### Explanation

Dictionary values can be of different data types, including integers, strings, floats, and Boolean values.

## 💻 Example 3: Empty Dictionary

````python
empty_dict = {}

print(empty_dict)
print(type(empty_dict))
````

### Explanation

An empty dictionary is created using `{}`.

## 💻 Example 4: Duplicate Keys

````python
student = {
    "name": "Nikita",
    "age": 20,
    "name": "Rahul"
}

print(student)
````

### Explanation

Dictionary keys must be unique.

If the same key appears more than once, the last value overwrites the previous one.

## 💻 Example 5: Duplicate Values

````python
students = {
    "student1": "Nikita",
    "student2": "Nikita",
    "student3": "Rahul"
}

print(students)
````

### Explanation

Values can be duplicated because only keys are required to be unique.

## 💻 Example 6: Length of a Dictionary

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(len(student))
````

### Explanation

The `len()` function returns the total number of key-value pairs in the dictionary.

## 💻 Example 7: Nested Dictionary

````python
students = {
    "student1": {
        "name": "Nikita",
        "marks": 92
    },
    "student2": {
        "name": "Rahul",
        "marks": 88
    }
}

print(students)
````

### Explanation

A nested dictionary stores another dictionary as a value.

It is useful for representing structured data such as student records or employee details.

## 🌍 Real-World Example: Employee Information

````python
employee = {
    "id": "EMP101",
    "name": "Nikita",
    "department": "Data Science",
    "salary": 65000
}

print(employee)
````

### Explanation

Employee records contain different types of related information.

Using a dictionary allows each value to be accessed quickly through a meaningful key such as `"name"` or `"salary"`.

# 🔑 Accessing Dictionary Values

Dictionary values are accessed using their **keys**.

Unlike lists and tuples, dictionaries do not use numeric indexes.

### Syntax

```python
dictionary[key]
```

## 💻 Example 1: Accessing Values Using Keys

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student["name"])
print(student["age"])
print(student["branch"])
````

### Explanation

Each value is retrieved using its corresponding key.

If the key exists, Python returns the associated value.

## ⚠️ Accessing a Missing Key

````python
student = {
    "name": "Nikita",
    "age": 20
}

print(student["marks"])
````

### Explanation

Trying to access a key that does not exist raises a **KeyError**.

## 🛠️ Using get()

The `get()` method safely retrieves a value.

If the key does not exist, it returns `None` (or a default value if provided) instead of raising an error.

### Syntax

```python
dictionary.get(key)
dictionary.get(key, default_value)
```

````python
student = {
    "name": "Nikita",
    "age": 20
}

print(student.get("name"))
print(student.get("marks"))
print(student.get("marks", "Not Available"))
````

### Explanation

`get()` is safer than square brackets because it avoids `KeyError`.

# ➕ Adding Key-Value Pairs

A new key-value pair can be added by assigning a value to a new key.

````python
student = {
    "name": "Nikita",
    "age": 20
}

student["branch"] = "CSE"

print(student)
````

### Explanation

Since `"branch"` does not already exist, Python adds it as a new key-value pair.

# 🔄 Updating Values

Existing values can be updated by assigning a new value to an existing key.

````python
student = {
    "name": "Nikita",
    "age": 20
}

student["age"] = 21

print(student)
````

### Explanation

If the key already exists, its value is replaced with the new value.

## 🔄 Using update()

The `update()` method adds or updates multiple key-value pairs at once.

### Syntax

```python
dictionary.update(other_dictionary)
```

````python
student = {
    "name": "Nikita",
    "age": 20
}

student.update({
    "age": 21,
    "branch": "CSE",
    "city": "Jaipur"
})

print(student)
````

### Explanation

- Existing keys are updated.
- New keys are added automatically.

# ❌ Removing Key-Value Pairs

## 🗑️ pop()

The `pop()` method removes a specified key and returns its value.

### Syntax

```python
dictionary.pop(key)
```

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

removed = student.pop("age")

print("Removed:", removed)
print(student)
````

### Explanation

`pop()` removes the specified key-value pair and returns the removed value.

## 🎲 popitem()

The `popitem()` method removes and returns the **last inserted key-value pair**.

### Syntax

```python
dictionary.popitem()
```

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student.popitem())
print(student)
````

### Explanation

In Python 3.7 and later, `popitem()` removes the **last inserted** key-value pair.

## 🧹 clear()

````python
student = {
    "name": "Nikita",
    "age": 20
}

student.clear()

print(student)
````

### Explanation

The `clear()` method removes all key-value pairs but keeps the dictionary object.

## 🗑️ del

The `del` keyword deletes a specific key or the entire dictionary.

````python
student = {
    "name": "Nikita",
    "age": 20
}

del student["age"]

print(student)
````

````python
student = {
    "name": "Nikita"
}

del student

print(student)
````

### Explanation

- `del dictionary[key]` removes a specific key-value pair.
- `del dictionary` deletes the entire dictionary from memory.

## 🌍 Real-World Example: Employee Database

````python
employee = {
    "id": "EMP101",
    "name": "Nikita",
    "department": "Data Science"
}

employee["salary"] = 65000
employee["department"] = "AI Engineer"

print(employee)
````

### Explanation

Employee information changes over time. Dictionaries make it easy to add new fields and update existing information using meaningful keys.

# 🛠️ Dictionary Methods

Dictionary methods help us retrieve, iterate, copy, and manipulate key-value pairs efficiently.

Some of the most commonly used methods are:

- `keys()`
- `values()`
- `items()`
- `copy()`
- `fromkeys()`
- `setdefault()`

## 🔑 1. keys()

The `keys()` method returns a view containing all the keys in the dictionary.

### Syntax

```python
dictionary.keys()
```

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student.keys())
````

### Explanation

The `keys()` method returns all keys without their corresponding values.

## 📄 2. values()

The `values()` method returns all values stored in the dictionary.

### Syntax

```python
dictionary.values()
```

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student.values())
````

### Explanation

The `values()` method returns only the values present in the dictionary.

## 🏷️ 3. items()

The `items()` method returns all key-value pairs as tuples.

### Syntax

```python
dictionary.items()
```

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

print(student.items())
````

### Explanation

Each item is returned as a tuple containing a key and its corresponding value.

# 🔁 Iterating Through a Dictionary

## 💻 Iterating Over Keys

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

for key in student:
    print(key)
````

### Explanation

By default, a `for` loop iterates through the dictionary keys.

## 💻 Iterating Over Values

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

for value in student.values():
    print(value)
````

### Explanation

Using `values()` allows iteration over all dictionary values.

## 💻 Iterating Over Keys and Values

````python
student = {
    "name": "Nikita",
    "age": 20,
    "branch": "CSE"
}

for key, value in student.items():
    print(key, ":", value)
````

### Explanation

The `items()` method makes it easy to access both keys and values during iteration.

## 📋 4. copy()

The `copy()` method creates a **shallow copy** of a dictionary.

### Syntax

```python
dictionary.copy()
```

````python
student = {
    "name": "Nikita",
    "age": 20
}

student_copy = student.copy()

print(student_copy)
````

### Explanation

Changes made to the copied dictionary do not affect the original dictionary (for simple data types).

## 🆕 5. fromkeys()

The `fromkeys()` method creates a new dictionary with the specified keys and a common default value.

### Syntax

```python
dict.fromkeys(keys, value)
```

````python
subjects = ["Python", "SQL", "Machine Learning"]

marks = dict.fromkeys(subjects, 0)

print(marks)
````

### Explanation

Each key is assigned the same default value.

## 🛠️ 6. setdefault()

The `setdefault()` method returns the value of a key.

If the key does not exist, it inserts the key with the specified default value.

### Syntax

```python
dictionary.setdefault(key, default_value)
```

````python
student = {
    "name": "Nikita",
    "age": 20
}

student.setdefault("branch", "CSE")

print(student)
````

### Explanation

Since `"branch"` does not exist, `setdefault()` adds it with the value `"CSE"`.

If the key already exists, its value remains unchanged.

# ⚡ Dictionary Comprehension

Dictionary comprehension provides a concise way to create dictionaries.

### Syntax

```python
{key: value for item in iterable}
```

````python
squares = {x: x**2 for x in range(1, 6)}

print(squares)
````

### Explanation

Dictionary comprehension creates dictionaries efficiently using a single line of code.

## 🌍 Real-World Example: Product Inventory

````python
inventory = {
    "Laptop": 15,
    "Mouse": 40,
    "Keyboard": 25
}

for product, quantity in inventory.items():
    print(product, ":", quantity)
````

### Explanation

Businesses often store inventory as dictionaries, where product names act as keys and stock quantities act as values. Iterating through the dictionary helps generate inventory reports efficiently.

## 📊 List vs Tuple vs Set vs Dictionary

| Feature           | List                     | Tuple                | Set            | Dictionary          |
| ----------------- | ------------------------ | -------------------- | -------------- | ------------------- |
| Syntax            | `[]`                     | `()`                 | `{}` / `set()` | `{key: value}`      |
| Ordered           | ✅ Yes                    | ✅ Yes                | ❌ No           | ✅ Yes (Python 3.7+) |
| Mutable           | ✅ Yes                    | ❌ No                 | ✅ Yes          | ✅ Yes               |
| Allows Duplicates | ✅ Yes                    | ✅ Yes                | ❌ No           | Keys ❌, Values ✅    |
| Indexing          | ✅ Yes                    | ✅ Yes                | ❌ No           | By Keys             |
| Slicing           | ✅ Yes                    | ✅ Yes                | ❌ No           | ❌ No                |
| Best Use Case     | Frequently changing data | Fixed/read-only data | Unique values  | Key-value mapping   |


## ⚠️ Common Mistakes

Beginners often make these mistakes while working with dictionaries:

- ❌ Trying to access a key that does not exist, resulting in a `KeyError`.
- ❌ Assuming dictionary keys can be duplicated (duplicate keys overwrite previous values).
- ❌ Confusing `keys()` with `values()`.
- ❌ Using mutable objects like lists as dictionary keys.
- ❌ Forgetting to use `items()` when iterating over both keys and values.
- ❌ Assuming dictionaries can be accessed using numeric indexes like lists.

## 💡 Best Practices

Follow these best practices while working with dictionaries:

- ✅ Use meaningful and descriptive key names.
- ✅ Use `get()` instead of square brackets when a key may not exist.
- ✅ Use `items()` when iterating over both keys and values.
- ✅ Keep keys unique and consistent.
- ✅ Use dictionary comprehensions for creating dictionaries efficiently.
- ✅ Use nested dictionaries to organize structured data.
