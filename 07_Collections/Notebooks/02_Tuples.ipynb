# 📦 Tuples in Python

## 📖 Introduction

A **tuple** is an ordered collection used to store multiple values in a single variable.

Tuples are similar to lists, but unlike lists, they are **immutable**, which means their elements cannot be added, removed, or modified after the tuple is created.

Tuples are useful when you want to store data that should remain constant throughout the program.

## 🤔 What is a Tuple?

A tuple is an **ordered, immutable collection** that allows duplicate values.

Tuples are created using parentheses `()`.

Since tuples cannot be modified after creation, they are often used for storing fixed data such as coordinates, dates, or configuration settings.

## ✨ Features of Tuples

Python tuples have the following features:

- Ordered collection
- Immutable (cannot be modified)
- Allows duplicate values
- Supports indexing and slicing
- Can store different data types
- Can contain nested tuples
- Faster than lists for read-only operations

## 💻 Example 1: Creating a Tuple

````python
fruits = ("Apple", "Banana", "Mango")

print(fruits)
````

### Explanation

The tuple contains three string elements enclosed in parentheses `()`.

Each element is separated by a comma.

## 💻 Example 2: Tuple with Different Data Types

````python
data = (101, "Nikita", 92.5, True)

print(data)
````

### Explanation

A tuple can store multiple data types together, including integers, strings, floating-point numbers, and Boolean values.

## 💻 Example 3: Empty Tuple

````python
empty_tuple = ()

print(empty_tuple)
print(type(empty_tuple))
````

### Explanation

An empty tuple contains no elements but is still a valid tuple object.

## 💻 Example 4: Single Element Tuple

To create a tuple with only one element, you **must** include a trailing comma.

Without the comma, Python treats it as a normal value enclosed in parentheses.

````python
student = ("Nikita",)

print(student)
print(type(student))
````

### Explanation

The comma after `"Nikita"` tells Python that this is a tuple containing one element.

## 💻 Example 5: Without Comma

````python
student = ("Nikita")

print(student)
print(type(student))
````

### Explanation

Since there is no trailing comma, Python treats `"Nikita"` as a string, not as a tuple.

## 💻 Example 6: Finding the Length of a Tuple

````python
numbers = (10, 20, 30, 40)

print(len(numbers))
````

### Explanation

The `len()` function returns the total number of elements in the tuple.

## 💻 Example 7: Nested Tuple

````python
students = (
    ("Nikita", 92),
    ("Rahul", 88),
    ("Priya", 95)
)

print(students)
````

### Explanation

A nested tuple is a tuple that contains one or more tuples as its elements.

Nested tuples are useful for storing structured and fixed data.

## 🌍 Real-World Example: GPS Coordinates

````python
location = (28.6139, 77.2090)

print(location)
````

### Explanation

GPS coordinates generally do not change, making a tuple an ideal data structure for storing latitude and longitude values.

# 🔢 Tuple Indexing

Like lists, tuples are **ordered collections**, so every element has a unique position called an **index**.

Python supports:

- **Positive Indexing**
- **Negative Indexing**

## 📍 Positive Indexing

Positive indexing starts from **0**.

```text
Index:      0        1        2        3
Tuple:   ("Apple","Banana","Mango","Orange")
```

````python
fruits = ("Apple", "Banana", "Mango", "Orange")

print(fruits[0])
print(fruits[1])
print(fruits[2])
print(fruits[3])
````

### Explanation

Each element is accessed using its index.

The first element always has index **0**.

## 📍 Negative Indexing

Negative indexing starts from the **last element**.

```text
Index:     -4      -3      -2      -1
Tuple:   ("Apple","Banana","Mango","Orange")
```

````python
fruits = ("Apple", "Banana", "Mango", "Orange")

print(fruits[-1])
print(fruits[-2])
print(fruits[-3])
print(fruits[-4])
````

### Explanation

Negative indexing is useful when you want to access elements from the end of the tuple.

## 💻 Example: Access Student Information

````python
student = ("Nikita", 20, "CSE", 92.5)

print("Name :", student[0])
print("Age :", student[1])
print("Branch :", student[2])
print("Marks :", student[3])
````

### Explanation

A tuple can store related information together, making it suitable for representing fixed records such as student information.

## ⚠️ IndexError

````python
fruits = ("Apple", "Banana", "Mango")

print(fruits[5])
````

### Explanation

Trying to access an index that does not exist raises an **IndexError**.

# ✂️ Tuple Slicing

Slicing allows you to extract a portion of a tuple.

### Syntax

```python
tuple[start : stop : step]
```

- **start** → Starting index (inclusive)
- **stop** → Ending index (exclusive)
- **step** → Number of positions to skip

## 💻 Example 1: Basic Slicing

````python
numbers = (10, 20, 30, 40, 50, 60)

print(numbers[1:4])
````

### Explanation

The slice starts from index **1** and stops before index **4**.

## 💻 Example 2: From Beginning

````python
numbers = (10, 20, 30, 40, 50)

print(numbers[:3])
````

### Explanation

If the starting index is omitted, Python starts from index **0**.

## 💻 Example 3: Till the End

````python
numbers = (10, 20, 30, 40, 50)

print(numbers[2:])
````

### Explanation

If the ending index is omitted, Python slices until the last element.

## 💻 Example 4: Step Slicing

````python
numbers = (10, 20, 30, 40, 50, 60)

print(numbers[::2])
````

### Explanation

The step value **2** selects every second element from the tuple.

## 💻 Example 5: Reverse a Tuple

````python
numbers = (10, 20, 30, 40, 50)

print(numbers[::-1])
````

### Explanation

A step value of **-1** reverses the tuple.

## 💻 Example 6: Accessing Nested Tuples

````python
students = (
    ("Nikita", 92),
    ("Rahul", 88),
    ("Priya", 95)
)

print(students[0])
print(students[0][0])
print(students[1][1])
````

### Explanation

Nested tuples allow you to organize structured data while keeping it immutable.

# 🔒 Tuple Immutability

One of the most important characteristics of tuples is **immutability**.

Once a tuple is created, its elements **cannot be changed, added, or removed**.

````python
fruits = ("Apple", "Banana", "Mango")

fruits[1] = "Orange"
````

### Explanation

Since tuples are immutable, attempting to modify an element raises a **TypeError**.

## 🌍 Real-World Example: RGB Color Code

````python
red = (255, 0, 0)

print(red)
````

### Explanation

RGB color values remain constant for a given color, making tuples a good choice for storing them.

# 🛠️ Tuple Methods

Tuples are immutable, so they have only **two built-in methods**:

- `count()`
- `index()`

## 1️⃣ count()

The `count()` method returns the number of times a specified value appears in the tuple.

### Syntax

```python
tuple.count(value)
```

````python
numbers = (10, 20, 10, 30, 10)

print(numbers.count(10))
````

### Explanation

The value `10` appears **three times**, so `count()` returns `3`.

## 2️⃣ index()

The `index()` method returns the index of the **first occurrence** of a specified value.

### Syntax

```python
tuple.index(value)
```

````python
fruits = ("Apple", "Banana", "Mango", "Banana")

print(fruits.index("Banana"))
````

### Explanation

Although `"Banana"` appears twice, `index()` returns the position of its **first occurrence**.

# 🔄 Tuple Operations

Tuples support several operators similar to lists.

Common operations include:

- Concatenation (`+`)
- Repetition (`*`)
- Membership (`in`)
- Non-membership (`not in`)

## 1️⃣ Concatenation (`+`)

````python
tuple1 = (1, 2, 3)
tuple2 = (4, 5, 6)

print(tuple1 + tuple2)
````

### Explanation

The `+` operator joins two tuples and creates a new tuple.

## 2️⃣ Repetition (`*`)

````python
numbers = (10, 20)

print(numbers * 3)
````

### Explanation

The `*` operator repeats the tuple the specified number of times.

## 3️⃣ Membership Operator (`in`)

````python
fruits = ("Apple", "Banana", "Mango")

print("Apple" in fruits)
print("Orange" in fruits)
````

### Explanation

The `in` operator checks whether an element exists in the tuple.

## 4️⃣ Non-Membership Operator (`not in`)

````python
fruits = ("Apple", "Banana", "Mango")

print("Orange" not in fruits)
````

### Explanation

The `not in` operator checks whether an element is absent from the tuple.

# 📊 Useful Built-in Functions

````python
numbers = (10, 20, 30, 40)

print("Length :", len(numbers))
print("Minimum :", min(numbers))
print("Maximum :", max(numbers))
print("Sum :", sum(numbers))
print("Sorted :", sorted(numbers))
````

### Explanation

These built-in functions work with tuples just like they do with lists.

**Note:** `sorted()` returns a **list**, not a tuple.

# 🎁 Tuple Packing

Tuple packing means storing multiple values into a single tuple.

````python
student = "Nikita", 20, "CSE"

print(student)
print(type(student))
````

### Explanation

Even without parentheses, Python automatically creates a tuple when multiple values are separated by commas.

# 🎁 Tuple Unpacking

Tuple unpacking assigns each element of a tuple to a separate variable.

````python
student = ("Nikita", 20, "CSE")

name, age, branch = student

print(name)
print(age)
print(branch)
````

### Explanation

Each tuple element is assigned to the corresponding variable in order.

## ⭐ Using `*` in Tuple Unpacking

````python
numbers = (10, 20, 30, 40, 50)

first, *middle, last = numbers

print(first)
print(middle)
print(last)
````

### Explanation

The `*` operator collects the remaining elements into a **list** during unpacking.

## 🌍 Real-World Example: Employee Record

````python
employee = ("EMP101", "Nikita", "Data Analyst")

emp_id, name, role = employee

print("Employee ID :", emp_id)
print("Name :", name)
print("Role :", role)
````

### Explanation

Employee records often contain fixed information that should not change, making tuples a suitable choice for storing such data.

| Slicing           | ✅ Yes           | ✅ Yes                        |
| Performance       | Slightly Slower | Slightly Faster              |
| Methods           | Many            | Only `count()` and `index()` |

| `min()`           | Return the smallest element             |

## ⚠️ Common Mistakes

Beginners often make these mistakes while working with tuples:

- ❌ Forgetting the comma when creating a single-element tuple.
- ❌ Trying to modify tuple elements.
- ❌ Using tuple methods that do not exist, such as `append()` or `remove()`.
- ❌ Confusing lists `[]` with tuples `()`.
- ❌ Assuming `sorted()` returns a tuple (it returns a list).
- ❌ Trying to unpack a tuple into the wrong number of variables.

## 💡 Best Practices

Follow these best practices while working with tuples:

- ✅ Use tuples for data that should not change.
- ✅ Use meaningful variable names.
- ✅ Use tuple unpacking to make your code more readable.
- ✅ Use lists instead of tuples if frequent modifications are required.
- ✅ Use tuples for fixed records such as coordinates, employee details, or configuration values.
- ✅ Remember to add a comma when creating a single-element tuple.
