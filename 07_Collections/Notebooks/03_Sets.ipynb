# 🎯 Sets in Python

## 📖 Introduction

A **set** is an unordered collection of unique elements.

Unlike lists and tuples, sets do not allow duplicate values and do not maintain the order of elements.

Sets are commonly used when you need to remove duplicate data, perform fast membership testing, or work with mathematical set operations such as union and intersection.

## 🤔 What is a Set?

A **set** is a built-in Python data type used to store multiple unique values.

A set:

- Stores only unique elements.
- Does not preserve insertion order.
- Does not support indexing.
- Is mutable, meaning elements can be added or removed.

## ✨ Features of Sets

Python sets have the following features:

- Unordered collection
- Mutable
- Stores only unique elements
- No duplicate values
- No indexing
- No slicing
- Supports mathematical set operations
- Fast membership testing

## 💻 Example 1: Creating a Set

````python
fruits = {"Apple", "Banana", "Mango"}

print(fruits)
````

### Explanation

A set is created using curly braces `{}`.

Unlike lists, sets automatically ensure that each element is unique.

## 💻 Example 2: Duplicate Values

````python
numbers = {10, 20, 30, 20, 10, 40}

print(numbers)
````

### Explanation

Duplicate values are automatically removed.

Only unique values remain in the set.

## 💻 Example 3: Empty Set

An empty set **cannot** be created using `{}`.

`{}` creates an empty dictionary.

To create an empty set, use the `set()` constructor.

````python
empty_set = set()

print(empty_set)
print(type(empty_set))
````

### Explanation

Always use `set()` when creating an empty set.

Using `{}` creates a dictionary instead.

## 💻 Example 4: Creating a Set from a List

````python
marks = [90, 85, 90, 95, 85]

unique_marks = set(marks)

print(unique_marks)
````

### Explanation

The `set()` function removes duplicate values from the list and keeps only unique elements.

## 💻 Example 5: Set with Different Data Types

````python
data = {101, "Nikita", 92.5, True}

print(data)
````

### Explanation

A set can store different data types together.

However, all elements must be **hashable (immutable)**.

## 💻 Example 6: Length of a Set

````python
colors = {"Red", "Green", "Blue"}

print(len(colors))
````

### Explanation

The `len()` function returns the total number of unique elements in the set.

## 💻 Example 7: Frozen Set

A **frozenset** is an immutable version of a set.

Once created, its elements cannot be changed.

````python
numbers = frozenset([10, 20, 30])

print(numbers)
print(type(numbers))
````

### Explanation

Use `frozenset()` when you need a read-only version of a set.

## 🌍 Real-World Example: Unique Website Visitors

````python
visitors = ["Amit", "Nikita", "Amit", "Rahul", "Nikita"]

unique_visitors = set(visitors)

print(unique_visitors)
````

### Explanation

A website may receive multiple visits from the same user.

Using a set helps count only **unique visitors**, making data analysis more accurate.

# ✏️ Modifying Sets

Unlike tuples, **sets are mutable**, which means you can add or remove elements after creating the set.

Python provides several built-in methods to modify sets.

## ➕ 1. add()

The `add()` method adds a **single element** to a set.

### Syntax

```python
set_name.add(element)
```

````python
fruits = {"Apple", "Banana", "Mango"}

fruits.add("Orange")

print(fruits)
````

### Explanation

The `add()` method inserts a single new element into the set.

If the element already exists, the set remains unchanged.

## ➕ 2. update()

The `update()` method adds **multiple elements** from another iterable.

### Syntax

```python
set_name.update(iterable)
```

````python
fruits = {"Apple", "Banana"}

fruits.update(["Mango", "Orange", "Grapes"])

print(fruits)
````

### Explanation

The `update()` method adds multiple elements at once.

Duplicate values are ignored automatically.

## ⚖️ add() vs update()

| Method     | Purpose                    |
| ---------- | -------------------------- |
| `add()`    | Adds **one element**       |
| `update()` | Adds **multiple elements** |


````python
numbers = {10, 20}

numbers.add(30)
numbers.update([40, 50])

print(numbers)
````

### Explanation

Use `add()` for a single value and `update()` when adding multiple values.

## ❌ 3. remove()

The `remove()` method deletes a specified element from the set.

### Syntax

```python
set_name.remove(element)
```

````python
fruits = {"Apple", "Banana", "Mango"}

fruits.remove("Banana")

print(fruits)
````

### Explanation

The specified element is removed from the set.

## ⚠️ remove() Error

````python
fruits = {"Apple", "Banana"}

fruits.remove("Orange")
````

### Explanation

If the element does not exist, `remove()` raises a **KeyError**.

## 🚮 4. discard()

The `discard()` method removes an element if it exists.

If the element is not found, no error is raised.

### Syntax

```python
set_name.discard(element)
```

````python
fruits = {"Apple", "Banana"}

fruits.discard("Orange")

print(fruits)
````

### Explanation

Unlike `remove()`, `discard()` safely ignores missing elements.

## ⚖️ remove() vs discard()

| Method      | Missing Element   |
| ----------- | ----------------- |
| `remove()`  | Raises `KeyError` |
| `discard()` | No Error          |


## 🎲 5. pop()

The `pop()` method removes and returns a **random element** from the set.

Since sets are unordered, you cannot predict which element will be removed.

````python
fruits = {"Apple", "Banana", "Mango"}

removed = fruits.pop()

print("Removed:", removed)
print("Remaining:", fruits)
````

### Explanation

The removed element may be different each time the program runs.

## 🧹 6. clear()

````python
fruits = {"Apple", "Banana", "Mango"}

fruits.clear()

print(fruits)
````

### Explanation

The `clear()` method removes all elements from the set but keeps the set object.

## 🗑️ 7. del

The `del` keyword deletes the entire set from memory.

````python
fruits = {"Apple", "Banana"}

del fruits

print(fruits)
````

### Explanation

After using `del`, the variable no longer exists, so accessing it raises a `NameError`.

## 🌍 Real-World Example: Online Course Enrollment

````python
students = {"Nikita", "Rahul"}

students.add("Priya")
students.discard("Rahul")

print(students)
````

### Explanation

A set is useful for storing enrolled students because each student appears only once. New students can be added, and students who leave the course can be removed easily.

# 🔄 Set Operations

Python sets support several mathematical operations.

The most commonly used operations are:

- 🤝 Union
- 🔍 Intersection
- ➖ Difference
- 🔄 Symmetric Difference
- ✅ Subset
- ✅ Superset
- 🚫 Disjoint

## 🤝 1. Union

The `union()` method returns a new set containing **all unique elements** from both sets.

### Syntax

```python
set1.union(set2)
```

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.union(set2))
````

### Explanation

The union combines both sets and automatically removes duplicate values.

## 💻 Union Using `|`

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1 | set2)
````

### Explanation

The `|` operator performs the same operation as `union()`.

## 🔍 2. Intersection

The `intersection()` method returns the elements that are **common to both sets**.

### Syntax

```python
set1.intersection(set2)
```

````python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print(set1.intersection(set2))
````

### Explanation

Only the elements present in both sets are returned.

## 💻 Intersection Using `&`

````python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}

print(set1 & set2)
````

### Explanation

The `&` operator performs the same operation as `intersection()`.

## ➖ 3. Difference

The `difference()` method returns elements that are present in the **first set but not in the second set**.

### Syntax

```python
set1.difference(set2)
```

````python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5}

print(set1.difference(set2))
````

### Explanation

Elements that exist only in the first set are returned.

## 💻 Difference Using `-`

````python
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5}

print(set1 - set2)
````

### Explanation

The `-` operator performs the same operation as `difference()`.

## 🔄 4. Symmetric Difference

The `symmetric_difference()` method returns elements that are present in **either set but not in both**.

### Syntax

```python
set1.symmetric_difference(set2)
```

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1.symmetric_difference(set2))
````

### Explanation

Common elements are removed, leaving only unique elements from each set.

## 💻 Symmetric Difference Using `^`

````python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

print(set1 ^ set2)
````

### Explanation

The `^` operator performs the same operation as `symmetric_difference()`.

## ✅ 5. issubset()

````python
set1 = {1, 2}
set2 = {1, 2, 3, 4}

print(set1.issubset(set2))
````

### Explanation

Returns `True` if every element of the first set exists in the second set.

## ✅ 6. issuperset()

````python
set1 = {1, 2, 3, 4}
set2 = {1, 2}

print(set1.issuperset(set2))
````

### Explanation

Returns `True` if the first set contains every element of the second set.

## 🚫 7. isdisjoint()

````python
set1 = {1, 2}
set2 = {3, 4}

print(set1.isdisjoint(set2))
````

### Explanation

Returns `True` if the two sets have **no common elements**.

## 🔍 Membership Operators

````python
fruits = {"Apple", "Banana", "Mango"}

print("Apple" in fruits)
print("Orange" not in fruits)
````

### Explanation

- `in` checks whether an element exists in the set.
- `not in` checks whether an element does not exist in the set.

| ------------------------ | --------------------------------------------------- |
| `add()`                  | Add a single element                                |
| `update()`               | Add multiple elements                               |
| `remove()`               | Remove an element (raises `KeyError` if missing)    |
| `discard()`              | Remove an element without raising an error          |
| `pop()`                  | Remove and return a random element                  |

| `union()`                | Return all unique elements from both sets           |

## 🌍 Real-World Example: Students Enrolled in Courses

````python
python_students = {"Nikita", "Rahul", "Priya"}
sql_students = {"Rahul", "Priya", "Aman"}

print("All Students:", python_students | sql_students)
print("Common Students:", python_students & sql_students)
print("Only Python Students:", python_students - sql_students)
print("Only One Course:", python_students ^ sql_students)
````

### Explanation

Educational platforms can use set operations to:

- Find students enrolled in both courses.
- Identify students enrolled in only one course.
- Create a complete list of enrolled students.
- Remove duplicate student names automatically.

| Best Use Case     | Frequently changing data | Fixed/read-only data | Unique values and set operations |


## ⚠️ Common Mistakes

Beginners often make these mistakes while working with sets:

- ❌ Using `{}` to create an empty set (it creates a dictionary).
- ❌ Expecting sets to maintain insertion order.
- ❌ Trying to access elements using indexing (`set[0]`).
- ❌ Adding mutable objects like lists to a set.
- ❌ Using `remove()` without checking if the element exists.
- ❌ Forgetting that duplicate values are automatically removed.

## 💡 Best Practices

Follow these best practices while working with sets:

- ✅ Use sets when you need unique values.
- ✅ Use `discard()` instead of `remove()` if the element may not exist.
- ✅ Use meaningful variable names.
- ✅ Use set operations (`union`, `intersection`, etc.) instead of writing unnecessary loops.
- ✅ Convert a list to a set to remove duplicate values quickly.
- ✅ Use `frozenset()` when you need an immutable set.
