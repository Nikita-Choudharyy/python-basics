# 📖 List Comprehension in Python

## 📖 Introduction

List Comprehension is a concise and elegant way to create lists in Python.

Instead of writing multiple lines of code using loops, List Comprehension allows you to create a list in a single line, making your code cleaner, shorter, and easier to read.

It is widely used in Data Science, Machine Learning, Data Analysis, and Python programming because it improves both readability and productivity.

## 🤔 What is List Comprehension?

List Comprehension is a compact way of creating a new list by processing elements from an iterable such as a list, tuple, string, or range.

Instead of writing a traditional `for` loop with `append()`, we can achieve the same result in a single expression.

### General Syntax

```python
[expression for item in iterable]
```

If a condition is required:

```python
[expression for item in iterable if condition]
```

## ✨ Why Use List Comprehension?

List Comprehension offers several advantages:

- Produces cleaner and shorter code.
- Improves readability.
- Often performs faster than using a traditional loop with `append()`.
- Makes data transformation simple.
- Widely used in Data Science and Machine Learning projects.

## 💻 Example 1: Creating a List Using a Loop

````python
numbers = []

for i in range(1, 6):
    numbers.append(i)

print(numbers)
````

### Explanation

A traditional `for` loop creates an empty list and adds elements one by one using `append()`.

## 💻 Example 2: Creating the Same List Using List Comprehension

````python
numbers = [i for i in range(1, 6)]

print(numbers)
````

### Explanation

The list is created in a single line.

This approach is shorter and easier to read.

## 💻 Example 3: Square of Numbers

````python
squares = [x**2 for x in range(1, 6)]

print(squares)
````

### Explanation

Each number is squared before being added to the new list.

## 💻 Example 4: Cube of Numbers

````python
cubes = [x**3 for x in range(1, 6)]

print(cubes)
````

### Explanation

The expression `x**3` is applied to every element before storing it in the list.

## 💻 Example 5: Convert Strings to Uppercase

````python
names = ["nikita", "rahul", "aman"]

upper_names = [name.upper() for name in names]

print(upper_names)
````

### Explanation

Each string is converted to uppercase before being stored in the new list.

## 💻 Example 6: Calculate String Length

````python
fruits = ["Apple", "Banana", "Mango"]

lengths = [len(fruit) for fruit in fruits]

print(lengths)
````

### Explanation

The `len()` function is applied to each string, and the lengths are stored in a new list.

## 💻 Example 7: Create a List from a String

````python
letters = [char for char in "Python"]

print(letters)
````

### Explanation

Strings are iterable, so each character is processed individually and stored in a list.

## 🌍 Real-World Example: Student Marks

````python
marks = [45, 60, 75, 82, 91]

updated_marks = [mark + 5 for mark in marks]

print(updated_marks)
````

### Explanation

Suppose every student receives a grace bonus of **5 marks**.

List Comprehension updates all marks efficiently in a single line without writing a loop.

# 🔍 Conditional List Comprehension

Sometimes, we want to include only specific elements in the new list.

This can be achieved by adding an **`if` condition** at the end of the list comprehension.

### Syntax

```python
[expression for item in iterable if condition]
```

## 💻 Example 1: Even Numbers

````python
even_numbers = [x for x in range(1, 11) if x % 2 == 0]

print(even_numbers)
````

### Explanation

Only numbers that satisfy the condition `x % 2 == 0` are added to the list.

## 💻 Example 2: Odd Numbers

````python
odd_numbers = [x for x in range(1, 11) if x % 2 != 0]

print(odd_numbers)
````

### Explanation

The condition selects only odd numbers from the given range.

## 💻 Example 3: Numbers Greater Than 5

````python
numbers = [x for x in range(1, 11) if x > 5]

print(numbers)
````

### Explanation

Only values greater than **5** satisfy the condition.

# 🔀 Using if-else in List Comprehension

Unlike a filtering condition, `if-else` allows us to transform every element based on a condition.

### Syntax

```python
[expression_if_true if condition else expression_if_false for item in iterable]
```

## 💻 Example 1: Even or Odd

````python
result = ["Even" if x % 2 == 0 else "Odd" for x in range(1, 11)]

print(result)
````

### Explanation

Each number is checked.

- If it is even → `"Even"` is added.
- Otherwise → `"Odd"` is added.

## 💻 Example 2: Pass or Fail

````python
marks = [35, 80, 45, 90, 28]

result = ["Pass" if mark >= 40 else "Fail" for mark in marks]

print(result)
````

### Explanation

Students scoring **40 or above** are marked as `"Pass"`.

Otherwise, they are marked as `"Fail"`.

# 🔁 Multiple for Loops

A List Comprehension can contain more than one `for` loop.

This is useful when generating combinations of values.

### Syntax

```python
[expression for item1 in iterable1 for item2 in iterable2]
```

## 💻 Example : Pair of Numbers

````python
pairs = [(x, y) for x in [1, 2, 3] for y in [10, 20]]

print(pairs)
````

### Explanation

For every value of `x`, Python iterates through all values of `y`, creating every possible pair.

# 🧩 Nested List Comprehension

Nested List Comprehension is used to process nested lists (lists inside lists).

It can also be used to flatten nested structures.

### Syntax

```python
[expression for row in matrix for item in row]
```

## 💻 Example 1: Flatten a Nested List

````python
matrix = [
    [1, 2],
    [3, 4],
    [5, 6]
]

flat = [num for row in matrix for num in row]

print(flat)
````

### Explanation

The outer loop selects each row, while the inner loop extracts every element from that row.

The result is a single flattened list.

## 💻 Example 2: First Letter of Each Word

````python
words = ["Python", "Machine", "Learning", "AI"]

letters = [word[0] for word in words]

print(letters)
````

### Explanation

The expression `word[0]` extracts the first character from every word in the list.

## 🌍 Real-World Example: Filter High Scores

````python
scores = [45, 90, 72, 38, 85, 60, 25]

high_scores = [score for score in scores if score >= 60]

print(high_scores)
````

### Explanation

Suppose a school wants to identify students who scored **60 or above**.

Instead of using a loop, List Comprehension filters the data in a single, readable line.

# ⚡ Advanced List Comprehension Examples

## 💻 Example 1: Using Functions

````python
numbers = [-4, -2, 0, 3, 7]

absolute_values = [abs(num) for num in numbers]

print(absolute_values)
````

### Explanation

The built-in `abs()` function is applied to every element in the list.

## 💻 Example 2: Remove Extra Spaces

````python
names = ["  Nikita  ", " Rahul ", " Aman "]

clean_names = [name.strip() for name in names]

print(clean_names)
````

### Explanation

The `strip()` method removes leading and trailing spaces from each string.

## 💻 Example 3: Filter Strings by Length

````python
words = ["AI", "Python", "ML", "Machine", "Data"]

long_words = [word for word in words if len(word) > 4]

print(long_words)
````

### Explanation

Only words with more than four characters are included in the new list.

## 💻 Example 4: Convert Numbers to Strings

````python
numbers = [10, 20, 30, 40]

strings = [str(num) for num in numbers]

print(strings)
````

### Explanation

Each integer is converted into a string using the `str()` function.

## 🌍 Real-World Example: Temperature Conversion

````python
celsius = [0, 20, 30, 40]

fahrenheit = [(temp * 9/5) + 32 for temp in celsius]

print(fahrenheit)
````

### Explanation

A weather application may store temperatures in Celsius but display them in Fahrenheit.

List Comprehension performs the conversion efficiently in one line.

## 📊 Quick Reference Table

| Syntax                                                               | Purpose                          |
| -------------------------------------------------------------------- | -------------------------------- |
| `[x for x in iterable]`                                              | Create a new list                |
| `[x for x in iterable if condition]`                                 | Filter elements                  |
| `[value_if_true if condition else value_if_false for x in iterable]` | Transform using `if-else`        |
| `[func(x) for x in iterable]`                                        | Apply a function to each element |
| `[item for row in matrix for item in row]`                           | Flatten a nested list            |


## 📊 Traditional Loop vs List Comprehension

| Feature     | Traditional Loop | List Comprehension                   |
| ----------- | ---------------- | ------------------------------------ |
| Code Length | Longer           | Shorter                              |
| Readability | Good             | Excellent (for simple tasks)         |
| Performance | Slightly slower  | Often faster                         |
| Best For    | Complex logic    | Simple transformations and filtering |


## ⚠️ Common Mistakes

Beginners often make these mistakes while using List Comprehension:

- ❌ Forgetting the correct order of `expression`, `for`, and `if`.
- ❌ Placing the `if` condition before the `for` loop.
- ❌ Confusing filtering (`if`) with transformation (`if-else`).
- ❌ Writing very long list comprehensions that reduce readability.
- ❌ Using List Comprehension when a normal loop is easier to understand.

## 💡 Best Practices

Follow these best practices while using List Comprehension:

- ✅ Keep list comprehensions short and readable.
- ✅ Use meaningful variable names.
- ✅ Prefer List Comprehension for simple transformations and filtering.
- ✅ Use normal loops for complex business logic.
- ✅ Avoid deeply nested list comprehensions unless necessary.
