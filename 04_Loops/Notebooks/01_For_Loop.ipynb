# 🔁 For Loop in Python

## 📖 Introduction

A **for loop** is used to execute a block of code repeatedly by iterating over a sequence such as a string, list, tuple, range, or other iterable objects.

It is one of the most commonly used looping statements in Python and is ideal when the number of iterations is known.

By the end of this notebook, you will be able to:

- Understand the purpose of a `for` loop.
- Write `for` loops using the `range()` function.
- Iterate through different sequences.
- Apply `for` loops to solve real-world problems.

## 🤔 What is a For Loop?

A **for loop** executes a block of code once for each item in a sequence.

Instead of writing the same code multiple times, a `for` loop automates repetitive tasks.

## 🌍 Real-World Examples

`for` loops are commonly used for:

- Processing student records
- Displaying product lists
- Reading data from files
- Iterating through datasets
- Training Machine Learning models

## 💻 Example: Iterating Over a List

```python
fruits = ["Apple", "Banana", "Mango"]

for fruit in fruits:
    print(fruit)
```

### Explanation

The loop visits each item in the list one by one and prints it.

The loop automatically stops after the last item is processed.

## 🔢 Using the `range()` Function

The `range()` function generates a sequence of numbers.

It is commonly used with `for` loops when you want to repeat a task a specific number of times.

```python
for number in range(5):
    print(number)
```

### Explanation

`range(5)` generates numbers from **0** to **4**.

The ending value (`5`) is **not included**.

## 📚 Different Forms of `range()`

The `range()` function can be used in three different ways:

| Syntax | Description |
|--------|-------------|
| `range(stop)` | Starts from `0` and ends before `stop`. |
| `range(start, stop)` | Starts from `start` and ends before `stop`. |
| `range(start, stop, step)` | Starts from `start`, ends before `stop`, and moves by `step`. |

### Example: `range(start, stop)`

```python
for number in range(1, 6):
    print(number)
```

### Explanation

The loop starts from **1** and continues until **5**.

The ending value (`6`) is not included.

### Example: `range(start, stop, step)`

```python
for number in range(2, 11, 2):
    print(number)
```

### Explanation

The loop starts from **2**, increases by **2** each time, and stops before **11**.

## 🔤 Iterating Over a String

```python
language = "Python"

for letter in language:
    print(letter)
```

### Explanation

Strings are sequences of characters.

The loop visits one character at a time.

## 🌍 Real-World Example

```python
students = ["Aman", "Nikita", "Riya", "Rahul"]

for student in students:
    print("Welcome", student)
```

### Explanation

The loop visits each student's name and displays a welcome message.

This approach avoids writing multiple `print()` statements manually.

```python
marks = [85, 92, 78, 95, 88]

total = 0

for mark in marks:
    total += mark

print("Total Marks:", total)
```

### Explanation

The loop processes each mark one by one and adds it to the `total` variable.

This is a common technique used to calculate totals, averages, and other statistics.

```python
prices = [250, 499, 799, 1200]

for price in prices:
    print("₹", price)
```

### Explanation

The loop visits each product price in the list and displays it one by one.

```python
attendance = ["Present", "Present", "Absent", "Present"]

for status in attendance:
    print(status)
```

### Explanation

The loop processes each student's attendance record one at a time.

```python
emails = [
    "nikita@gmail.com",
    "rahul@gmail.com",
    "riya@gmail.com"
]

for email in emails:
    print("Sending email to:", email)
```

### Explanation

The loop sends (simulates sending) an email to each recipient in the list.

```python
tasks = [
    "Study Python",
    "Practice Coding",
    "Complete Assignment"
]

for task in tasks:
    print("✔", task)
```

### Explanation

The loop displays each task from the to-do list one by one.

```python
cities = ["Jaipur", "Delhi", "Mumbai", "Bengaluru"]

for city in cities:
    print("Delivering Order to:", city)
```

### Explanation

The loop processes each city where an order needs to be delivered.

```python
scores = [78, 91, 86, 95]

for score in scores:
    if score >= 90:
        print(score, "- Excellent")
    else:
        print(score, "- Good")
```

### Explanation

The loop checks each student's score and displays a message based on the marks obtained.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting the colon (`:`) after the `for` statement.
- ❌ Incorrect indentation inside the loop.
- ❌ Expecting `range(5)` to include `5`.
- ❌ Modifying the loop variable unnecessarily inside the loop.

## 💡 Best Practices

Follow these best practices while writing `for` loops:

- ✅ Use meaningful loop variable names.
- ✅ Keep the loop body simple and readable.
- ✅ Choose the appropriate form of `range()`.
- ✅ Avoid unnecessary nested loops when a single loop is sufficient.
