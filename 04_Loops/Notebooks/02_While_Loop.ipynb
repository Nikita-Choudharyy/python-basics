# 🔄 While Loop in Python

## 📖 Introduction

A **while loop** repeatedly executes a block of code **as long as a specified condition remains `True`**.

Unlike a `for` loop, which is commonly used when the number of iterations is known, a `while` loop is useful when the number of iterations depends on a condition.

## 🤔 What is a While Loop?

A **while loop** executes a block of code repeatedly **while a condition is `True`**.

Before each iteration, Python evaluates the condition.

- If the condition is `True`, the loop continues.
- If the condition is `False`, the loop stops.

## 🌍 Real-World Examples

`while` loops are commonly used in:

- Login systems
- ATM transactions
- Menu-driven programs
- Games
- Waiting for user input
- Reading data until a condition is met

## 💻 Example: Print Numbers 

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

### Explanation

The loop starts with `count = 1`.

After each iteration, the value of `count` increases by `1`.

When `count` becomes `6`, the condition `count <= 5` becomes `False`, and the loop stops.

## 💻 Example: Print Numbers in Reverse

```python
num = 5

while num >= 1:
    print(num)
    num -= 1
```

### Explanation

The loop starts from **5** and decreases the value by **1** after each iteration.

It stops when the value becomes **0**.

## 💻 Example: Print Even Numbers

```python
num = 2

while num <= 10:
    print(num)
    num += 2
```

### Explanation

The loop starts from **2** and increases the value by **2** in each iteration, printing only even numbers.

## 💻 Example: Multiplication Table

```python
number = 5
i = 1

while i <= 10:
    print(number * i)
    i += 1
```

### Explanation

The loop runs **10 times** and prints the multiplication table of **5**.

## 💻 Example: Sum of First Five Numbers

```python
num = 1
total = 0

while num <= 5:
    total += num
    num += 1

print("Sum:", total)
```

### Explanation

The loop adds each number to the `total` variable until the value reaches **5**.

## 💻 Example: Print Squares of Numbers

```python
num = 1

while num <= 5:
    print(num ** 2)
    num += 1
```

### Explanation

The loop prints the square of each number from **1** to **5**.

## ⚠️ Infinite Loop

An **infinite loop** occurs when the loop condition never becomes `False`.

This usually happens when the loop variable is not updated.

```python
count = 1

while count <= 5:
    print(count)
```

> **⚠️ Do not run this program.**

Since the value of `count` never changes, the condition always remains `True`, causing the loop to run forever.

## ✅ Fixing the Infinite Loop

```python
count = 1

while count <= 5:
    print(count)
    count += 1
```

### Explanation

Updating the loop variable ensures that the condition eventually becomes `False`, allowing the loop to terminate normally.

## 🌍 Real-World Examples

```python
password = ""

while password != "python123":
    password = input("Enter Password: ")

print("Login Successful")
```

### Explanation

The loop continues asking the user to enter the password until the correct password is entered.

Once the correct password is provided, the loop stops.

```python
balance = 5000

while balance > 0:
    print("Remaining Balance:", balance)
    balance -= 1000
```

### Explanation

The loop keeps running while the account balance is greater than zero.

After each iteration, ₹1,000 is deducted from the balance until it reaches zero.

```python
choice = "yes"

while choice == "yes":
    print("Program is Running")
    choice = input("Do you want to continue? (yes/no): ")
```

### Explanation

The program continues to run as long as the user enters **"yes"**.

When the user enters **"no"**, the condition becomes `False`, and the loop ends.

```python
temperature = 40

while temperature > 30:
    print("Current Temperature:", temperature, "°C")
    temperature -= 2

print("Temperature is now comfortable.")
```

### Explanation

The loop continues until the temperature drops to **30°C or below**.

After each iteration, the temperature decreases by **2°C**.

```python
battery = 100

while battery > 20:
    print("Battery Level:", battery, "%")
    battery -= 15

print("Low Battery! Please charge your device.")
```

### Explanation

The battery level decreases after each iteration.

The loop stops once the battery level reaches **20% or below**.

```python
stock = 50

while stock > 0:
    print("Items Remaining:", stock)
    stock -= 10

print("Out of Stock")
```

### Explanation

The loop simulates selling products.

After every sale, the stock decreases until no items remain.

```python
remaining_time = 5

while remaining_time > 0:
    print("Exam Time Left:", remaining_time, "minutes")
    remaining_time -= 1

print("Time is Over!")
```

### Explanation

The loop displays the remaining exam time.

After each minute passes, the remaining time decreases until the exam ends.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting to update the loop variable.
- ❌ Creating infinite loops unintentionally.
- ❌ Using incorrect loop conditions.
- ❌ Incorrect indentation inside the loop.

## 💡 Best Practices

Follow these best practices while writing `while` loops:

- ✅ Ensure the loop condition eventually becomes `False`.
- ✅ Update the loop variable correctly.
- ✅ Keep the loop body simple and readable.
- ✅ Use meaningful variable names.
