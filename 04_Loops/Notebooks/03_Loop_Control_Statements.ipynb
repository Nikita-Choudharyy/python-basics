# ⏯️ Loop Control Statements in Python

## 📖 Introduction

Loop control statements change the normal execution of a loop.

They allow you to:

- Stop a loop before it finishes.
- Skip specific iterations.
- Reserve a block of code for future implementation.

Python provides three loop control statements:

- `break`
- `continue`
- `pass`

## 🎯 Learning Objectives

By the end of this notebook, you will be able to:

- Understand the purpose of loop control statements.
- Use the `break` statement.
- Use the `continue` statement.
- Use the `pass` statement.
- Apply loop control statements in practical programs.

## 🛑 The `break` Statement

The `break` statement immediately terminates the loop.

As soon as Python encounters `break`, it exits the loop and continues executing the code after the loop.

## 💻 Example 1: Stop the Loop at a Specific Number

```python
for number in range(1, 11):
    if number == 6:
        break

    print(number)
```

### Explanation

The loop prints numbers from **1** to **5**.

When the value becomes **6**, the `break` statement terminates the loop immediately.

## 💻 Example 2: Stop When a Negative Number is Found

```python
numbers = [12, 25, 18, -7, 30, 45]

for num in numbers:
    if num < 0:
        break

    print(num)
```

### Explanation

The loop processes each number in the list.

As soon as a negative number is found, the loop terminates using `break`.

## 💻 Example 3: Stop When a Character is Found

```python
word = "PYTHON"

for letter in word:
    if letter == "H":
        break

    print(letter)
```

### Explanation

The loop prints each character until it encounters the letter **H**.

When **H** is found, the loop stops immediately.

## 💻 Example 4: Stop at the First Multiple of 7

```python
for number in range(10, 31):
    if number % 7 == 0:
        break

    print(number)
```

### Explanation

The loop starts from **10**.

When it reaches the first multiple of **7** (`14`), the `break` statement stops the loop.

## 💻 Example 5: Stop Searching in a List

```python
fruits = ["Apple", "Banana", "Mango", "Orange"]

for fruit in fruits:
    if fruit == "Mango":
        print("Fruit Found")
        break

    print("Checking:", fruit)
```

### Explanation

The loop checks each fruit one by one.

When **Mango** is found, it prints **"Fruit Found"** and immediately exits the loop.

## 🌍 Real-World Example

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor"]

search_item = "Keyboard"

for product in products:
    if product == search_item:
        print("Product Found")
        break

    print("Checking:", product)
```

### Explanation

The loop searches for a product in the list.

Once the required product is found, the `break` statement stops the search.

```python
correct_pin = 1234

attempts = [1111, 2222, 1234, 5678]

for pin in attempts:
    if pin == correct_pin:
        print("Login Successful")
        break

    print("Incorrect PIN")
```

### Explanation

The program checks each login attempt.

Once the correct PIN is entered, the user is logged in and the loop stops.

```python
inventory = ["Pen", "Notebook", "Marker", "Eraser"]

required_item = "Marker"

for item in inventory:
    if item == required_item:
        print("Item Available")
        break

    print("Searching:", item)
```

### Explanation

The loop searches for the required item in the inventory.

When the item is found, the `break` statement exits the loop without checking the remaining items.

## ⏭️ The `continue` Statement

The `continue` statement skips the current iteration and immediately moves to the next iteration of the loop.

Unlike `break`, it does **not** terminate the loop.

### Example 1: Skip a Specific Number

```python
for number in range(1, 6):
    if number == 3:
        continue

    print(number)
```

### Explanation

When the value becomes **3**, the `continue` statement skips that iteration.

The loop then continues with the next value.

### Example 2: Skip Even Numbers

```python
for number in range(1, 11):
    if number % 2 == 0:
        continue

    print(number)
```

### Explanation

The loop skips all even numbers and prints only odd numbers.

### Example 3: Skip Vowels

```python
word = "PYTHON"

for letter in word:
    if letter in "AEIOU":
        continue

    print(letter)
```

### Explanation

Whenever a vowel is found, the `continue` statement skips it.

Only consonants are printed.

### Example 4: Skip Negative Numbers

```python
numbers = [10, -5, 20, -8, 30]

for num in numbers:
    if num < 0:
        continue

    print(num)
```

### Explanation

Negative numbers are skipped, while positive numbers are processed normally.

### Example 5: Skip Multiples of 3

```python
for number in range(1, 16):
    if number % 3 == 0:
        continue

    print(number)
```

### Explanation

Whenever the number is divisible by **3**, that iteration is skipped.

The remaining numbers are printed.

## 🌍 Real-World Examples

```python
students = [
    "Aman",
    "Absent",
    "Nikita",
    "Absent",
    "Riya"
]

for student in students:
    if student == "Absent":
        continue

    print("Present:", student)
```

### Explanation

The loop skips students marked as **Absent** and processes only those who are present.

```python
products = [
    "Laptop",
    "Out of Stock",
    "Keyboard",
    "Mouse"
]

for product in products:
    if product == "Out of Stock":
        continue

    print("Available:", product)
```

### Explanation

Products that are out of stock are skipped, while available products are displayed.

```python
transactions = [
    "Success",
    "Failed",
    "Success",
    "Failed",
    "Success"
]

for transaction in transactions:
    if transaction == "Failed":
        continue

    print("Processing:", transaction)
```

### Explanation

The loop ignores failed transactions and processes only successful ones.

## ⏸️ The `pass` Statement

The `pass` statement is a **placeholder**.

It does nothing when executed and is commonly used when a statement is required syntactically, but no action needs to be performed.

### Example 1: Empty `if` Block

```python
age = 18

if age >= 18:
    pass

print("Program Finished")
```

### Explanation

The `pass` statement does nothing.

It allows the program to continue executing without producing an error.

### Example 2: Empty Loop

```python
for number in range(5):
    pass

print("Loop Completed")
```

### Explanation

The loop executes normally, but `pass` performs no action during each iteration.

### Example 3: Empty Function

```python
def greet():
    pass

print("Function Created")
```

### Explanation

The function is created successfully.

The `pass` statement acts as a placeholder until the function is implemented.

## 🌍 Real-World Example: Feature Under Development

```python
feature_ready = False

if not feature_ready:
    pass
else:
    print("Launching Feature")

print("Application Started")
```

### Explanation

The feature is not ready yet, so `pass` acts as a placeholder.

The program continues running without performing any action.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Confusing `break` with `continue`.
- ❌ Using `break` when they only want to skip one iteration.
- ❌ Forgetting that `continue` does not stop the loop.
- ❌ Assuming `pass` skips an iteration like `continue`.
- ❌ Using `pass` when actual logic is required.

## 💡 Best Practices

Follow these best practices while using loop control statements:

- ✅ Use `break` to exit a loop when no further iterations are needed.
- ✅ Use `continue` to skip only the current iteration.
- ✅ Use `pass` as a temporary placeholder during development.
- ✅ Choose the statement that best matches the required behavior.
