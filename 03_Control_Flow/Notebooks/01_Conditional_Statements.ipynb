# 🔀 Conditional Statements in Python

## 📖 Introduction

Programs don't always execute every line of code. In many situations, they need to make decisions based on specific conditions.

For example:

- Is a student eligible to vote?
- Did a user enter the correct password?
- Is a product in stock?
- Has a student passed the exam?

Conditional statements allow Python to execute different blocks of code depending on whether a condition is **True** or **False**.

## 🤔 What are Conditional Statements?

Conditional statements are used to make decisions in a program.

A condition is evaluated first.

- If the condition is **True**, one block of code executes.
- If the condition is **False**, another block of code may execute.

They help programs respond differently based on different situations.

## 🌍 Real-World Examples

Conditional statements are used in many real-life applications, such as:

- ATM withdrawal validation
- Login authentication
- Student result processing
- Online shopping discounts
- Age verification
- Traffic signal systems

## 📋 Types of Conditional Statements

Python provides three commonly used conditional statements:

| Statement | Purpose |
|-----------|---------|
| `if` | Executes code only when a condition is `True`. |
| `if-else` | Chooses between two blocks of code. |
| `if-elif-else` | Checks multiple conditions. |

# ✅ The `if` Statement

The `if` statement executes a block of code only when the specified condition is **True**.

### Example

```python
age = 20

if age >= 18:
    print("You are eligible to vote.")
```

### Explanation

The condition `age >= 18` evaluates to **True**.

Therefore, Python executes the code inside the `if` block.

## 💡 Indentation in Python

Python uses **indentation (spaces)** to define blocks of code.

The code inside an `if` statement must be indented.

Incorrect indentation will result in an `IndentationError`.

```python
marks = 85

if marks >= 40:
    print("Pass")

print("Program Finished")
```

In the above example:

- The first `print()` statement is inside the `if` block.
- The second `print()` statement is outside the `if` block, so it executes regardless of the condition.

# 🔄 The `if-else` Statement

The `if-else` statement is used when you want to execute one block of code if the condition is **True** and another block if the condition is **False**.

### Example

```python
age = 16

if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

### Explanation

The condition `age >= 18` evaluates to **False**.

Therefore, Python skips the `if` block and executes the `else` block.

## 🌍 Real-World Example

```python
password = "python123"

if password == "python123":
    print("Login Successful")
else:
    print("Invalid Password")
```

The program checks whether the entered password matches the correct password.

If the condition is `True`, the user is allowed to log in; otherwise, an error message is displayed.

# 🔀 The `if-elif-else` Statement

The `if-elif-else` statement is used when there are **multiple conditions** to evaluate.

Python checks each condition one by one.

As soon as a condition becomes `True`, its corresponding block is executed, and the remaining conditions are skipped.

### Example

```python
marks = 82

if marks >= 90:
    print("Grade A")
elif marks >= 75:
    print("Grade B")
elif marks >= 60:
    print("Grade C")
else:
    print("Grade D")
```

### Explanation

The conditions are checked from top to bottom.

- `marks >= 90` → False
- `marks >= 75` → True ✅

Since the second condition is `True`, Python prints **Grade B** and does not check the remaining conditions.

## 🌍 Real-World Example

```python
amount = 7500

if amount >= 10000:
    discount = 20
elif amount >= 5000:
    discount = 10
else:
    discount = 5

print("Discount:", discount, "%")
```

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting the colon (`:`) after `if`, `elif`, or `else`.
- ❌ Incorrect indentation.
- ❌ Using `=` instead of `==` for comparison.
- ❌ Writing conditions in the wrong order in an `if-elif-else` statement.

## 💡 Best Practices

Follow these best practices while writing conditional statements:

- ✅ Keep conditions simple and readable.
- ✅ Use meaningful variable names.
- ✅ Write the most specific conditions before more general ones.
- ✅ Maintain proper indentation for better readability.
