# 🪆 Nested Conditions in Python

## 📖 Introduction

Sometimes a single condition is not enough to solve a problem.

In such situations, we place one conditional statement inside another. This is called a **Nested Condition**.

Nested conditions allow programs to make more detailed decisions based on multiple levels of checking.

## 🎯 Learning Objectives

By the end of this notebook, you will be able to:

- Understand nested conditions.
- Write nested `if` statements.
- Solve real-world problems using nested conditions.
- Improve the readability of nested code.

## 🤔 What are Nested Conditions?

A nested condition is a conditional statement placed inside another conditional statement.

The inner condition is checked **only if the outer condition is satisfied**.

This allows programs to perform more specific checks before executing code.

## 💻 Example 1

```python
age = 22
has_license = True

if age >= 18:
    if has_license:
        print("You can drive.")
```

### Explanation

The program first checks whether the person is **18 years or older**.

Only if this condition is `True` does Python check whether the person has a valid driving license.

Since both conditions are satisfied, the message **"You can drive."** is displayed.

## 💻 Example 2

```python
percentage = 92
attendance = 80

if percentage >= 90:
    if attendance >= 75:
        print("Scholarship Approved")
```

The outer condition checks the student's percentage.

Only if it is **90 or above** does Python check the attendance requirement.

## 💻 Example 3

```python
pin_verified = True
balance = 2500

if pin_verified:
    if balance >= 1000:
        print("Withdrawal Successful")
    else:
        print("Insufficient Balance")
else:
    print("Invalid PIN")
```

### Explanation

The program first verifies the PIN.

If the PIN is correct, it then checks the account balance.

The withdrawal is successful only when **both conditions** are satisfied.

## 💻 Example 4

```python
department = "IT"
experience = 6

if department == "IT":
    if experience >= 10:
        print("Bonus: ₹50,000")
    elif experience >= 5:
        print("Bonus: ₹25,000")
    else:
        print("Bonus: ₹10,000")
else:
    print("Bonus Policy Not Available")
```

Python first checks the employee's department.

If the employee belongs to the IT department, it then checks the years of experience to determine the bonus amount.

## 🌍 Real-World Example

```python
course_completed = True
certificate_generated = True

if course_completed:
    if certificate_generated:
        print("Certificate Ready for Download")
    else:
        print("Certificate is Being Generated")
else:
    print("Complete the Course First")
```

### Explanation

The program first checks whether the course is completed.

If it is completed, Python then checks whether the certificate has been generated.

```python
username = "nikita"
password = "python123"
otp_verified = True

if username == "nikita":
    if password == "python123":
        if otp_verified:
            print("Login Successful")
        else:
            print("OTP Verification Failed")
    else:
        print("Incorrect Password")
else:
    print("Invalid Username")
```

### Explanation

The program verifies the username, password, and OTP one by one.

The user is logged in only when all three conditions are satisfied.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using unnecessary nested conditions.
- ❌ Incorrect indentation.
- ❌ Forgetting that the inner condition runs only if the outer condition is `True`.
- ❌ Writing deeply nested code that is difficult to read.

## 💡 Best Practices

Follow these best practices while writing nested conditions:

- ✅ Keep nesting as simple as possible.
- ✅ Use meaningful variable names.
- ✅ Maintain proper indentation.
- ✅ Use logical operators (`and`, `or`) when they simplify the code.
