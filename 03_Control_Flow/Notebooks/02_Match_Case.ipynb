# 🔀 Match-Case Statement in Python

## 📖 Introduction

The **match-case** statement is used to compare a value against multiple possible cases.

It provides a clean and readable alternative when checking many specific values.

The `match-case` statement was introduced in **Python 3.10**.

## 🎯 Learning Objectives

By the end of this notebook, you will be able to:

- Understand the purpose of the `match-case` statement.
- Write programs using `match-case`.
- Use the default (`case _`) option.
- Apply `match-case` to real-world scenarios.

## 🤔 What is Match-Case?

The `match-case` statement compares a single value against multiple cases.

Python checks each case one by one.

When a matching case is found, the corresponding block of code is executed.

If no case matches, the default case (`_`) can be used.

## 💻 Example

```python
day = 3

match day:
    case 1:
        print("Monday")
    case 2:
        print("Tuesday")
    case 3:
        print("Wednesday")
    case 4:
        print("Thursday")
    case 5:
        print("Friday")
    case _:
        print("Invalid Day")
```

### Explanation

The value of `day` is `3`.

Python checks each case until it finds a match.

Since `case 3` matches, it prints **Wednesday** and stops checking the remaining cases.

## ⭐ The Default Case (`_`)

The underscore (`_`) acts as the **default case**.

It executes when none of the specified cases match the given value.

```python
month = 15

match month:
    case 1:
        print("January")
    case 2:
        print("February")
    case _:
        print("Invalid Month")
```

### Explanation

Since the value `15` does not match any defined case, the default case (`_`) is executed.

## 🌍 Real-World Example

```python
role = "admin"

match role:
    case "admin":
        print("Full Access")
    case "editor":
        print("Edit Access")
    case "viewer":
        print("Read Only Access")
    case _:
        print("Access Denied")
```

### Explanation

The value of `role` is `"admin"`.

Python finds the matching case and executes its corresponding block.

Since a match is found, the remaining cases are skipped.

## 💡 When Should You Use `match-case`?

Use `match-case` when:

- You are comparing **one variable** against multiple fixed values.
- The program has many possible options.
- You want your code to be clean and easy to read.

Examples include:

- Menu-driven programs
- Day or month selection
- User roles
- Product categories
- Command processing

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Using `match-case` in Python versions earlier than **3.10**.
- ❌ Forgetting to include the default case (`case _`) when needed.
- ❌ Trying to replace every `if-else` statement with `match-case`.
- ❌ Using incorrect indentation inside the `case` blocks.

## 💡 Best Practices

Follow these best practices:

- ✅ Use `match-case` when comparing a single variable with multiple fixed values.
- ✅ Include a default case (`case _`) whenever appropriate.
- ✅ Keep each case short and focused.
- ✅ Use meaningful variable names.
