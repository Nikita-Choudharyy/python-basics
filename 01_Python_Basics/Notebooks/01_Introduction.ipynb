📜 History of Python

Python was created by Guido van Rossum in 1989 and was officially released in 1991.

The language was designed with the goal of making programming more readable and productive.

Interestingly, the name Python was inspired by the British comedy television show "Monty Python's Flying Circus", not by the snake.

Over time, Python has evolved significantly, with Python 3 becoming the standard version used today.

## 🧠 How Python Executes Code

When you run a Python program, it goes through several steps before producing the final output.

The execution process is shown below:

```text
Python Source Code (.py)
          │
          ▼
   Python Interpreter
          │
          ▼
   Bytecode (.pyc)
          │
          ▼
Python Virtual Machine (PVM)
          │
          ▼
        Output
```

### Explanation

- **Python Source Code:** The code written by the programmer.
- **Python Interpreter:** Converts the source code into bytecode.
- **Bytecode:** An intermediate form of the program.
- **Python Virtual Machine (PVM):** Executes the bytecode.
- **Output:** The final result displayed on the screen.

## ⚠️ Common Mistakes

While learning Python, beginners often make the following mistakes:

- ❌ Forgetting to install Python correctly.
- ❌ Not adding Python to the system PATH during installation.
- ❌ Writing `Print()` instead of `print()`.
- ❌ Forgetting quotation marks around strings.
- ❌ Not saving the file before running it.
- ❌ Confusing the Terminal with the Python Interpreter.
