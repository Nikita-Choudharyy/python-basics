# 🔄 Type Casting in Python

## 📖 Introduction

In Python, different operations require different data types. Sometimes, you need to convert a value from one data type to another.

This process is called **Type Casting** or **Type Conversion**.

For example, when taking input from the user, Python stores the input as a string. To perform mathematical operations, you must convert it into an integer or a float.

## 🧠 What is Type Casting?

Type Casting is the process of converting one data type into another.

Python provides two types of type casting:

- **Implicit Type Casting** (Automatic Conversion)
- **Explicit Type Casting** (Manual Conversion)

## 🤔 Why Do We Need Type Casting?

Type casting is useful because it allows us to:

- Perform mathematical operations.
- Convert user input into the required data type.
- Improve compatibility between different data types.
- Prevent type-related errors.

## 🔄 Types of Type Casting

### 1. Implicit Type Casting

Python automatically converts one data type into another whenever it is safe to do so.

```python
num1 = 10
num2 = 5.5

result = num1 + num2

print(result)
print(type(result))
```

### Explanation

The integer (`int`) value is automatically converted into a floating-point (`float`) value before the addition.

This automatic conversion is known as **Implicit Type Casting**.

### 2. Explicit Type Casting

In explicit type casting, the programmer manually converts a value from one data type to another using built-in functions.

Common type casting functions:

- `int()`
- `float()`
- `str()`
- `bool()`
- `list()`
- `tuple()`
- `set()`

```python
num = "25"

converted_num = int(num)

print(converted_num)
print(type(converted_num))
```

## 🔢 Converting to Integer using `int()`

The `int()` function converts a compatible value into an integer.

It removes the decimal part of floating-point numbers and converts numeric strings into integers.

```python
num = 15.8

print(int(num))
```

```python
age = "21"

print(int(age))
```

### 💡 Note

The `int()` function **does not round** a decimal number.

It simply removes the decimal part (truncates the value).

## 💰 Converting to Float using `float()`
The `float()` function converts integers and numeric strings into floating-point numbers.

```python
marks = 95

print(float(marks))
```

```python
price = "499.99"

print(float(price))
```

## 📝 Converting to String using `str()`
The `str()` function converts different data types into strings.

This is useful when displaying numbers as text or combining numbers with strings.

```python
roll_number = 101

print(str(roll_number))
print(type(str(roll_number)))
```

## ✅ Converting to Boolean using `bool()`
The `bool()` function converts values into Boolean (`True` or `False`).

Generally:

- Empty values → `False`
- Non-empty values → `True`

```python
print(bool(1))
print(bool(0))

print(bool("Python"))
print(bool(""))
```

### Common Boolean Conversions

| Value | Result |
|-------|--------|
| `0` | `False` |
| `1` | `True` |
| `""` | `False` |
| `"Python"` | `True` |
| `[]` | `False` |
| `[1, 2]` | `True` |
| `None` | `False` |

## 🔄 Real-World Example

```python
price = input("Enter product price: ")

price = float(price)

gst = 50

total_price = price + gst

print("Total Price:", total_price)
```

### Explanation

The `input()` function returns the entered value as a string.

Using `float()`, we convert it into a numeric value so mathematical operations can be performed.

## ⚠️ Common Mistakes

Beginners often make the following mistakes while performing type casting:

- ❌ Trying to convert a non-numeric string to an integer.
- ❌ Forgetting to convert user input before performing calculations.
- ❌ Assuming `int()` rounds decimal values.
- ❌ Mixing different data types without conversion.
- ❌ Confusing implicit and explicit type casting.

### Example: Invalid Type Conversion
value = "Python"

print(int(value))

Running the above code will raise a **ValueError** because `"Python"` cannot be converted into an integer.

## 💡 Best Practices

Follow these best practices while working with type casting:

- ✅ Convert data only when required.
- ✅ Check the input before converting it.
- ✅ Use meaningful variable names.
- ✅ Use `int()` for whole numbers and `float()` for decimal numbers.
- ✅ Read error messages carefully if a conversion fails.
