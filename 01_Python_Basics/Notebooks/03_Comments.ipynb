## 🧠 What are Comments?

Comments are notes written inside a program to explain the code.

Python ignores comments during program execution.

They are mainly used to improve code readability and make collaboration easier.

## ❓ Why Do We Use Comments?

Comments are useful because they:

- Explain complex code.
- Improve code readability.
- Help during debugging.
- Make code easier to maintain.
- Allow developers to leave notes or reminders.
- Improve collaboration in team projects.

## 📝 Single-Line Comments

A single-line comment begins with the `#` symbol.

Everything written after `#` on the same line is treated as a comment and ignored by Python.

```python
# This is a single-line comment

print("Hello, World!")
```

### Explanation

The first line is a comment and is ignored by Python.

Only the `print()` statement is executed.

## 📝 Comments After Code

```python
print("Python")   # Display the word Python
```

### Explanation

Comments can also be written at the end of a line of code to explain what that specific line does.

## 📝 Multiple Single-Line Comments

```python
# Student Details
# Name: Nikita
# Course: Python Basics

print("Welcome!")
```

### Explanation

Although each line starts with `#`, together they form a block of comments used to describe the program.

## 📝 Multi-Line Comments

Python does not have a dedicated syntax for multi-line comments.

The most common way is to write multiple single-line comments using the `#` symbol.

This approach is recommended because it is clear and widely used by Python developers.

```python
# This program demonstrates
# how multiple single-line comments
# can be used together.

print("Hello, Python!")
```

### Explanation

Each line starts with the `#` symbol, making it a comment.

Python ignores all comment lines and executes only the `print()` statement.

## 📄 Docstrings

Docstrings (Documentation Strings) are special strings enclosed within triple quotes (`""" """` or `''' '''`).

They are primarily used to document modules, functions, classes, and methods.

Unlike regular comments, docstrings can be accessed using Python's built-in `help()` function or the `__doc__` attribute.

```python
def greet():
    """
    This function prints
    a welcome message.
    """
    print("Welcome!")

help(greet)
```

### Explanation

The text enclosed within triple quotes is the function's **docstring**.

It provides documentation about the purpose of the function and can be viewed using `help()`.

## 🔍 Comments vs Docstrings

| Comments | Docstrings |
|----------|------------|
| Used to explain code. | Used to document modules, classes, and functions. |
| Ignored by Python during execution. | Stored as an object's documentation. |
| Start with `#`. | Written using triple quotes (`""" """` or `''' '''`). |
| Cannot be accessed programmatically. | Can be accessed using `help()` or `__doc__`. |

## 🧠 When Should You Write Comments?

Write comments when:

- Explaining complex logic.
- Clarifying the purpose of a code block.
- Providing important notes or warnings.
- Documenting assumptions or limitations.

Avoid writing comments for code that is already obvious.

## ❌ Poor Comments vs ✅ Good Comments



```python
# Bad Comment
x = 10  # Assign 10 to x
```

```python
# Good Comment
# Maximum number of login attempts allowed
max_attempts = 10
```

### Explanation

A good comment provides useful information that is not immediately obvious from the code.

Avoid writing comments that simply repeat what the code already makes clear.

## 📖 Code Without Comments vs Code With Comments
The following examples demonstrate how comments improve code readability and make it easier to understand the purpose of the program.

```python
# Without Comments

radius = 7
area = 3.14 * radius * radius

print(area)
```

```python
# With Comments

# Radius of the circle
radius = 7

# Calculate the area using the formula πr²
area = 3.14 * radius * radius

# Display the calculated area
print(area)
```

### Explanation

Both programs produce the same output.

However, the second program is much easier to understand because the comments explain the purpose of each step.

## 💡 Best Practices

Follow these best practices while writing comments:

- ✅ Write comments only when necessary.
- ✅ Keep comments short and meaningful.
- ✅ Update comments whenever the code changes.
- ✅ Use comments to explain **why**, not **what**.
- ✅ Use docstrings to document functions, classes, and modules.

## ⚠️ Common Mistakes

Beginners often make these mistakes while writing comments:

- ❌ Writing unnecessary comments for obvious code.
- ❌ Leaving outdated comments after modifying code.
- ❌ Writing very long comments that reduce readability.
- ❌ Using comments instead of meaningful variable names.
- ❌ Confusing comments with docstrings.
