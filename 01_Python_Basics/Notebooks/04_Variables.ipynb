# 📦 Variables in Python

## 📖 Introduction

Variables are one of the most fundamental concepts in Python programming.

A variable acts as a container that stores data in memory. Instead of using values directly throughout a program, we store them in variables so they can be reused, modified, and managed efficiently.

Whether you're building a simple calculator or a Machine Learning model, variables are used everywhere in Python.

## 🧠 What is a Variable?

A **variable** is a named memory location used to store data.

Instead of remembering actual values, we use variable names to access and manipulate the stored data.

Think of a variable as a labeled container where information can be stored and retrieved whenever needed.

## 🌍 Why Do We Need Variables?

Variables make programs easier to write, read, and maintain.

They allow us to:

- Store data for later use.
- Reuse values multiple times.
- Update values whenever needed.
- Make programs dynamic and interactive.
- Improve code readability.

## 📦 Creating Variables

In Python, variables are created automatically when you assign a value to them.

Unlike many programming languages, you do **not** need to declare the variable type before using it.

### Syntax

variable_name = value

### Example 1: Creating Variables

````python
name = "Nikita"
age = 21
height = 5.4

print(name)
print(age)
print(height)
````

### Explanation

- `name` stores a string.
- `age` stores an integer.
- `height` stores a floating-point number.
- The `print()` function displays the stored values.

## 🔄 Reassigning Variables

Variables can be updated at any time by assigning a new value to them.

The new value replaces the old one.

````python
city = "Jaipur"

print(city)

city = "Delhi"

print(city)
````

### Explanation

Initially, the variable `city` stores `"Jaipur"`.

Later, it is reassigned to `"Delhi"`, replacing the previous value.

## 📦 Assigning Multiple Variables

Python allows multiple variables to be assigned in a single statement.

````python
x, y, z = 10, 20, 30

print(x)
print(y)
print(z)
````

### Example: Assigning the Same Value

````python
a = b = c = 100

print(a)
print(b)
print(c)
````

### Explanation

Python allows you to assign the same value to multiple variables using a single assignment statement.

## 🧠 Memory Tip

Think of a variable as a **labeled storage box**.

- 📦 The **box** stores the value.
- 🏷️ The **label** is the variable name.

Instead of remembering the actual value, Python remembers the variable name, which points to that value.

## 🏷️ Variable Naming Rules

Python follows certain rules while naming variables.

A valid variable name:

- ✅ Must begin with a letter (`A-Z` or `a-z`) or an underscore (`_`).
- ✅ Can contain letters, digits, and underscores.
- ✅ Is case-sensitive.
- ✅ Cannot contain spaces.
- ✅ Cannot begin with a number.
- ✅ Cannot use Python keywords.

### ✅ Valid Variable Names

````python
name = "Nikita"

_age = 21

student_name = "Rahul"

marks2026 = 95
````

### ❌ Invalid Variable Names

2name = "Nikita"

student-name = "Rahul"

my name = "Amit"

class = "Python"

### Explanation

The above variable names are invalid because:

- `2name` starts with a number.
- `student-name` contains a hyphen (`-`).
- `my name` contains a space.
- `class` is a reserved Python keyword.

## 🔠 Variable Naming Conventions

Although Python allows many naming styles, developers generally follow these conventions:

- **snake_case** → Recommended for variable names.
- **camelCase** → Common in some programming languages.
- **PascalCase** → Mostly used for class names.
- **UPPER_CASE** → Used for constants.

````python
# snake_case
student_name = "Nikita"

# camelCase
studentName = "Nikita"

# PascalCase
StudentName = "Nikita"

# UPPER_CASE
PI = 3.14159
````

### 💡 Best Practice

Follow **snake_case** while naming variables because it follows Python's official style guide (PEP 8).

## 🔄 Dynamic Typing

Python is a **dynamically typed** programming language.

This means you do not need to specify the data type while creating a variable.

Python automatically determines the type based on the assigned value.

````python
x = 10

print(type(x))

x = "Python"

print(type(x))
````

### Explanation

Initially, `x` stores an integer.

Later, the same variable stores a string.

Python automatically changes the variable's type based on the assigned value.

## 📌 Multiple Assignment Examples

````python
name, age, city = "Nikita", 21, "Jaipur"

print(name)
print(age)
print(city)
````

````python
x = y = z = "Python"

print(x)
print(y)
print(z)
````

### Explanation

Python allows:

- Assigning different values to multiple variables.
- Assigning the same value to multiple variables.

Both approaches make code shorter and more readable.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Starting a variable name with a number.
- ❌ Using spaces in variable names.
- ❌ Using Python keywords as variable names.
- ❌ Creating unclear variable names such as `a`, `b`, and `c` for large programs.
- ❌ Confusing `Name` and `name` (Python is case-sensitive).

## 💡 Best Practices

Follow these best practices while naming variables:

- ✅ Use meaningful names.
- ✅ Follow the snake_case naming convention.
- ✅ Keep names short but descriptive.
- ✅ Avoid unnecessary abbreviations.
- ✅ Use lowercase letters for variables.
