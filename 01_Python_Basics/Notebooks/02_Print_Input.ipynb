## 🧠 The `print()` Function

The `print()` function is a built-in Python function used to display output on the screen.

It can print:

- Strings
- Numbers
- Variables
- Expressions
- Multiple values

### Syntax

````python
print(object)
````

### Example 1: Printing a String

````python
print("Hello, World!")
````

### Explanation

- `print()` displays the given value on the screen.
- `"Hello, World!"` is a string enclosed in quotation marks.

### Example 2: Printing Numbers

````python
print(10)
print(3.14)
````

### Example 3: Printing Multiple Values

````python
print("Age:",25)
print("Height:",5.8)
````

### Example 4: Using `sep`

````python
print("2026","07","08",sep="-")
````

### Explanation

The `sep` parameter specifies the separator inserted between multiple values.

Default separator:

```
(space)
```

### Example 5: Using `end`

````python
print("Hello",end=" ")
print("Python")
````

### Explanation

By default, `print()` ends with a new line (`\n`).

The `end` parameter allows you to change that behavior.

## 🧠 The `input()` Function

The `input()` function is used to receive input from the user while the program is running.

The entered value is always returned as a **string**.

### Syntax

````python
input("Enter something: ")
````

### Example 1: Taking User Input

````python
name = input("Enter your name: ")

print("Hello,",name)
````

### Example 2: Taking Multiple Inputs

````python
name = input("Enter your name: ")
age = input("Enter your age: ")

print(name)
print(age)
````

### Example 3: Converting Input to Integer

````python
age = int(input("Enter your age: "))

print(age)
````

### Why Use `int()`?

The `input()` function always returns data as a string.

To perform mathematical operations, we convert the input into an integer using the `int()` function.

## ⚠️ Common Mistakes

Beginners often make the following mistakes while using `print()` and `input()`:

- ❌ Forgetting quotation marks around strings.
- ❌ Using `Print()` instead of `print()`.
- ❌ Forgetting parentheses.
- ❌ Assuming `input()` returns an integer.
- ❌ Performing arithmetic without converting the input using `int()` or `float()`.

### Example: Incorrect Code

````python
age = input("Enter your age: ")

print(age + 10)
````

The above code will raise a **TypeError** because `age` is a string, while `10` is an integer.

To perform arithmetic operations, convert the input using `int()`.

### Correct Code

````python
age = int(input("Enter your age: "))

print(age + 10)
````
