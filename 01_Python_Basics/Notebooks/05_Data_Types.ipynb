# 📊 Data Types in Python

## 📖 Introduction

Every value stored in a Python program has a specific **data type**.

A data type tells Python what kind of value is being stored and determines the operations that can be performed on that value.

For example, you can add two numbers together, but you cannot directly add a number and a string without converting one of them.

Understanding data types is essential because they form the foundation of Python programming.

## 🧠 What is a Data Type?

A **data type** defines the kind of value stored in a variable.

Python uses data types to determine:

- How the value is stored in memory.
- Which operations can be performed on it.
- How Python interprets the value.

Every value in Python belongs to a specific data type.

## 🤔 Why Do We Need Data Types?

Data types are important because they help Python:

- Store data efficiently.
- Perform valid operations.
- Prevent invalid operations.
- Improve program reliability.
- Manage memory effectively.

## 🔍 The `type()` Function

The `type()` function is used to determine the data type of any object in Python.

It returns the class (type) of the given object.

### Syntax

```python
type(object)
```

### Example 1: Integer

```python
age = 21

print(type(age))
```

### Explanation

The variable `age` stores an integer value.

Therefore, `type(age)` returns `<class 'int'>`.

### Example 2: Float

```python
price = 99.99

print(type(price))
```

### Example 3: String

```python
course = "Python"

print(type(course))
```

### Example 4: Boolean

```python
is_student = True

print(type(is_student))
```

## 🧩 Numeric Data Types

Python provides three built-in numeric data types:

- **int** → Integer numbers
- **float** → Floating-point (decimal) numbers
- **complex** → Complex numbers

### Integer (`int`)
Integers are whole numbers without a decimal point.

They can be positive, negative, or zero.

```python
age = 21
temperature = -5
score = 0

print(age)
print(temperature)
print(score)
```

### Floating-Point (`float`)
Floating-point numbers (floats) contain decimal values.

They are commonly used for measurements, prices, percentages, and scientific calculations.

```python
price = 99.99
height = 5.8
pi = 3.14159

print(price)
print(height)
print(pi)
```

### Complex (`complex`)
Complex numbers contain two parts:

- A real part.
- An imaginary part represented by `j`.

They are mainly used in scientific computing, engineering, and mathematics.

```python
number = 3 + 4j

print(number)
print(type(number))
```

## 🧾 Text Data Type (`str`)
A string (`str`) is used to store textual data.

Strings can be enclosed using:

- Single quotes (`' '`)
- Double quotes (`" "`)
- Triple quotes (`''' '''` or `""" """`)

```python
name = "Nikita"
city = 'Jaipur'

message = """Welcome to
Python Programming"""

print(name)
print(city)
print(message)
```

## ✅ Boolean Data Type (`bool`)
A Boolean data type has only two possible values:

- `True`
- `False`

Boolean values are widely used in decision-making and conditional statements.

```python
is_logged_in = True
is_admin = False

print(is_logged_in)
print(is_admin)
```

## 🚫 None Data Type

The `None` value represents the absence of a value.

It is commonly used to indicate that a variable currently has no meaningful data.

```python
result = None

print(result)
print(type(result))
```

## 🔍 The `isinstance()` Function
The `isinstance()` function is used to check whether an object belongs to a specific data type.

It returns either:

- `True`
- `False`

### Syntax
isinstance(object, data_type)

```python
age = 21

print(isinstance(age, int))
print(isinstance(age, float))
```

### Explanation

Since `age` is an integer, `isinstance(age, int)` returns `True`.

It returns `False` when checked against the `float` data type.

## 🔄 Mutable vs Immutable Data Types

Python data types are classified into two categories:

- **Mutable** → Their values can be changed after creation.
- **Immutable** → Their values cannot be changed after creation.

### Mutable Data Types

Mutable objects allow their contents to be modified without creating a new object.

Examples include:

- `list`
- `dict`
- `set`
- `bytearray`

```python
fruits = ["Apple", "Banana"]

fruits.append("Mango")

print(fruits)
```

### Immutable Data Types
Immutable objects cannot be modified after they are created.

Examples include:

- `int`
- `float`
- `bool`
- `str`
- `tuple`
- `frozenset`
- `bytes`

```python
name = "Python"

print(name)
```

Although the value of an immutable object cannot be changed, you can assign a completely new value to the variable.

```python
name = "Python"

name = "Java"

print(name)
```

### 🧠 Memory Tip

- **Mutable** = Can Change ✅
- **Immutable** = Cannot Change ❌

A simple way to remember:

> **Mutable → Modify**
>
> **Immutable → Impossible to Modify**
