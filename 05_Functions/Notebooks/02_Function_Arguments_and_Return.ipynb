## 🤔 Parameters vs Arguments

Although the terms **parameter** and **argument** are often used together, they have different meanings.

- **Parameter:** A variable defined in the function definition.
- **Argument:** The actual value passed when calling the function.

### Example

```python
def greet(name):      # name → Parameter
    print(f"Hello, {name}!")

greet("Nikita")       # "Nikita" → Argument
```

### Explanation

The variable **`name`** is called a **parameter** because it is defined inside the function.

The value **"Nikita"** is called an **argument** because it is passed while calling the function.

## 📝 Function Syntax with Parameters

```python
'''
def function_name(parameter):
     Function Body

function_name(argument)
'''
```

## 💻 Example 1: Single Parameter

```python
def greet(name):
    print(f"Welcome, {name}!")

greet("Aman")
```

### Explanation

The function accepts one parameter and displays a personalized welcome message.

## 💻 Example 2: Multiple Parameters

```python
def student(name, course):
    print(f"Name : {name}")
    print(f"Course : {course}")

student("Nikita", "Python")
```

### Explanation

The function accepts two parameters and prints both values.

## 💻 Example 3: Function with Three Parameters

```python
def employee(name, department, city):
    print(f"Name : {name}")
    print(f"Department : {department}")
    print(f"City : {city}")

employee("Rahul", "HR", "Jaipur")
```

### Explanation

A function can accept multiple parameters depending on the requirement.

## 📍 Positional Arguments

In **positional arguments**, values are assigned to parameters based on their position.

The order of arguments must match the order of parameters.

### Example 1: Positional Arguments

```python
def student(name, course):
    print(f"Name   : {name}")
    print(f"Course : {course}")

student("Nikita", "Python")
```

### Explanation

The first argument is assigned to `name` and the second argument is assigned to `course`.

The order of arguments is important.

### Example 2: Changing the Order

```python
def student(name, course):
    print(f"Name   : {name}")
    print(f"Course : {course}")

student("Python", "Nikita")
```

### Explanation

Python assigns values based only on their position.

Since the arguments were swapped, the output is also incorrect.

## 🏷️ Keyword Arguments

In **keyword arguments**, parameter names are specified while calling the function.

The order of arguments does not matter.

### Example 1: Keyword Arguments

```python
def student(name, course):
    print(f"Name   : {name}")
    print(f"Course : {course}")

student(course="Python", name="Nikita")
```

### Explanation

The values are assigned using parameter names.

The order does not matter because Python matches arguments by keyword.

### Example 2: Three Keyword Arguments

```python
def employee(name, department, city):
    print(f"Name       : {name}")
    print(f"Department : {department}")
    print(f"City       : {city}")

employee(
    city="Jaipur",
    name="Rahul",
    department="HR"
)
```

### Explanation

Even though the order is different, the correct values are assigned because keyword arguments are used.

## 🎯 Default Arguments

A **default argument** provides a default value for a parameter.

If no value is passed during the function call, the default value is used.

### Example 1: Using a Default Value

```python
def greet(name="Guest"):
    print(f"Welcome, {name}!")

greet()
```

### Explanation

No argument is passed, so Python uses the default value `"Guest"`.

### Example 2: Overriding the Default Value

```python
def greet(name="Guest"):
    print(f"Welcome, {name}!")

greet("Nikita")
```

### Explanation

Since an argument is provided, it replaces the default value.

## 🌍 Real-World Example: Online Food Order

### Explanation

The customer provides the food item and quantity while placing an order.

These values are passed as arguments to the function.

The number of seats defaults to **1** if the customer doesn't specify a value.

## 🌍 Real-World Example: Employee Details

## ⭐ Variable-Length Arguments (`*args`)

Sometimes we don't know how many arguments will be passed to a function.

In such cases, we use **`*args`**.

It allows a function to accept any number of positional arguments.

### Syntax

```python
def function_name(*args):
    # Function Body
    pass 
```

### Example 1

```python
def numbers(*args):
    print(args)

numbers(10, 20, 30, 40)
```

### Explanation

All positional arguments are collected into a **tuple**.

The number of arguments can vary each time the function is called.

### Example 2

```python
def total(*numbers):
    print(sum(numbers))

total(10, 20)
total(10, 20, 30)
total(10, 20, 30, 40)
```

### Explanation

The same function works with different numbers of arguments without any changes.

## ⭐ Keyword Variable-Length Arguments (`**kwargs`)

`**kwargs` allows a function to accept any number of keyword arguments.

All keyword arguments are stored inside a **dictionary**.

### Example 1

```python
def student(**kwargs):
    print(kwargs)

student(name="Nikita", course="Python", city="Jaipur")
```

### Explanation

All keyword arguments are stored as **key-value pairs** inside a dictionary.

### Example 2

```python
def employee(**details):
    for key, value in details.items():
        print(f"{key} : {value}")

employee(
    name="Rahul",
    department="HR",
    city="Delhi"
)
```

### Explanation

The `.items()` method is used to access both keys and values from the dictionary.

## 🔙 The `return` Statement

The `return` statement sends a value back to the place where the function was called.

Unlike `print()`, the returned value can be stored, reused, or used in further calculations.

### Example 1: Returning a Value

```python
def square(number):
    return number ** 2

result = square(5)

print(result)
```

### Explanation

The function calculates the square and returns the result.

The returned value is stored in the variable `result`.

### Example 2: Returning Multiple Values

```python
def calculation(a, b):
    return a + b, a - b

addition, subtraction = calculation(20, 10)

print(addition)
print(subtraction)
```

### Explanation

Python allows multiple values to be returned.

The returned values are automatically packed into a tuple and can be unpacked into separate variables.

## 🆚 `print()` vs `return`

| Feature                     | `print()` | `return` |
| --------------------------- | --------- | -------- |
| Displays output             | ✅         | ❌        |
| Sends value back            | ❌         | ✅        |
| Can be stored in a variable | ❌         | ✅        |
| Can be reused later         | ❌         | ✅        |
| Used inside functions       | ✅         | ✅        |


```python
def using_print():
    print("Hello")

def using_return():
    return "Hello"

value = using_return()

using_print()
print(value)
```

### Explanation

`using_print()` only displays the message.

`using_return()` sends the value back, allowing it to be stored in a variable and used later.

## 🌍 Real-World Example: Student Result

```python
def calculate_percentage(total_marks, obtained_marks):
    percentage = (obtained_marks / total_marks) * 100
    return percentage

result = calculate_percentage(500, 425)

print(f"Percentage: {result}%")
```

### Explanation

The function calculates the percentage and returns it.

The returned value is stored and then displayed.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Confusing parameters with arguments.
- ❌ Passing arguments in the wrong order when using positional arguments.
- ❌ Forgetting to return a value when it is needed.
- ❌ Using `print()` instead of `return`.
- ❌ Forgetting to use `*` with `args` or `**` with `kwargs`.
- ❌ Passing fewer or more arguments than required.

## 💡 Best Practices

Follow these best practices while writing functions:

- ✅ Use meaningful parameter names.
- ✅ Use keyword arguments when a function has many parameters.
- ✅ Use default arguments for optional values.
- ✅ Return values instead of printing whenever possible.
- ✅ Use `*args` and `**kwargs` only when flexibility is required.
