# ⚡ Lambda Functions

## 🤔 What is a Lambda Function?

A **Lambda Function** is a small anonymous function created using the `lambda` keyword.

Unlike normal functions, a lambda function:

- Does not require the `def` keyword.
- Does not have a function name.
- Can contain only one expression.
- Automatically returns the result of that expression.

## 📝 Syntax

lambda parameters: expression

- `lambda` → Keyword used to create an anonymous function.
- `parameters` → Input values.
- `expression` → The operation to perform.

## 💻 Example 1: Square of a Number

```python
square = lambda x: x ** 2

print(square(5))
```

### Explanation

The lambda function takes one parameter and returns its square.

## 💻 Example 2: Addition

```python
add = lambda a, b: a + b

print(add(10, 20))
```

### Explanation

The lambda function accepts two numbers and returns their sum.

## 💻 Example 3: Even or Odd

```python
even_or_odd = lambda number: "Even" if number % 2 == 0 else "Odd"

print(even_or_odd(12))
print(even_or_odd(9))
```

### Explanation

The lambda function uses a conditional expression to determine whether a number is even or odd.

## 🌍 Real-World Example: Student Marks

```python
students = [
    ("Aman", 85),
    ("Nikita", 92),
    ("Riya", 78)
]

students.sort(key=lambda student: student[1])

print(students)
```

### Explanation

The lambda function tells Python to sort the list using the marks instead of the student names.

# 🛠️ Common Built-in Functions

Python includes many built-in functions.

Below are some of the most commonly used ones for beginners and data science learners.

## 1️⃣ len()

```python
fruits = ["Apple", "Banana", "Mango"]

print(len(fruits))
```

## 2️⃣ type()

```python
print(type(100))
print(type("Python"))
```

## 3️⃣ sum()

```python
numbers = [10,20,30]

print(sum(numbers))
```

## 4️⃣ max() and min()

```python
numbers = [12,8,45,21]

print(max(numbers))
print(min(numbers))
```

## 5️⃣ sorted()

```python
numbers = [5,2,8,1]

print(sorted(numbers))
```

## 6️⃣ abs()

```python
print(abs(-25))
```

## 7️⃣ round()

```python
print(round(3.14159,2))
```

## 8️⃣ enumerate()

```python
fruits = ["Apple","Banana","Mango"]

for index, fruit in enumerate(fruits):
    print(index, fruit)
```

## 9️⃣ zip()

```python
names = ["Aman","Nikita"]

marks = [80,95]

for name, mark in zip(names, marks):
    print(name, mark)
```

## 🔟 any() and all()

```python
print(any([False, False, True]))

print(all([True, True, False]))
```

## ⚠️ Common Mistakes

 - ❌ Using lambda for complex logic.
 - ❌ Forgetting that lambda contains only one expression.
 - ❌ Trying to write multiple statements inside a lambda.
 - ❌ Using built-in function names (list, sum, max) as variable names.

## 💡 Best Practices

 - ✅ Use lambda for short operations.
 - ✅ Use normal functions for complex logic.
 - ✅ Learn commonly used built-in functions first.
 - ✅ Write readable code instead of overly clever code.
