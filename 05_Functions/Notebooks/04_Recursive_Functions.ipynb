# 🔄 Recursive Functions in Python

## 📖 Introduction

A **recursive function** is a function that calls itself to solve a problem.

Instead of using loops, recursion breaks a large problem into smaller, similar problems until a stopping condition is reached.

Recursion is commonly used in algorithms, tree structures, graph traversal, and many programming problems.

## 🤔 What is Recursion?

A **recursive function** is a function that calls itself.

Every recursive function must have:

- **Base Case** → Stops the recursion.
- **Recursive Case** → Calls the function again with a smaller or simpler problem.

Without a base case, the function would continue calling itself indefinitely, resulting in a **RecursionError**.

## 📝 General Syntax

def function_name(value):

    if base_condition:
        return result

    return function_name(smaller_value)

## 🧠 How Recursion Works

Suppose we call:

```python
countdown(3)
```

The function calls happen like this:

```text
countdown(3)
    ↓
countdown(2)
    ↓
countdown(1)
    ↓
countdown(0)
    ↓
Base Case Reached
```

Once the base case is reached, the function stops calling itself and returns.

## 💻 Example 1: Countdown

````python
def countdown(number):

    if number == 0:
        print("Done!")
        return

    print(number)

    countdown(number - 1)

countdown(5)
````

### Explanation

The function prints the current number.

It then calls itself with a smaller value.

When the number becomes **0**, the base case is reached and recursion stops.

## 💻 Example 2: Count Up

````python
def count_up(number):

    if number == 0:
        return

    count_up(number - 1)

    print(number)

count_up(5)
````

### Explanation

The recursive calls continue until the base case is reached.

The numbers are printed while the function calls return, resulting in ascending order.

## 💻 Example 3: Sum of Natural Numbers

````python
def natural_sum(number):

    if number == 1:
        return 1

    return number + natural_sum(number - 1)

result = natural_sum(5)

print(result)
````

### Explanation

The function keeps adding the current number to the sum of smaller numbers.

For `natural_sum(5)`:

```text
5 + 4 + 3 + 2 + 1 = 15
```

## 💻 Example 4: Factorial

````python
def factorial(number):

    if number == 0 or number == 1:
        return 1

    return number * factorial(number - 1)

print(factorial(5))
````

### Explanation

Each recursive call multiplies the current number by the factorial of the previous number.

```text
5 × 4 × 3 × 2 × 1 = 120
```

## 💻 Example 5: Fibonacci Sequence

````python
def fibonacci(number):

    if number <= 1:
        return number

    return fibonacci(number - 1) + fibonacci(number - 2)

for i in range(8):
    print(fibonacci(i), end=" ")
````

### Explanation

Each Fibonacci number is the sum of the previous two numbers.

The recursive function continues until it reaches the base cases `0` and `1`.

## 💻 Example 6: Reverse a String

````python
def reverse_string(text):

    if len(text) == 0:
        return ""

    return text[-1] + reverse_string(text[:-1])

print(reverse_string("Python"))
````

### Explanation

The function takes the last character of the string and recursively reverses the remaining part.

The recursion continues until an empty string is reached.

## 💻 Example 7: Power of a Number

````python
def power(base, exponent):

    if exponent == 0:
        return 1

    return base * power(base, exponent - 1)

print(power(2, 5))
````

### Explanation

The function repeatedly multiplies the base until the exponent becomes **0**.

For `2⁵`, the calculation is:

```text
2 × 2 × 2 × 2 × 2 = 32
```

## 💻 Example 8: Count Digits

````python
def count_digits(number):

    if number < 10:
        return 1

    return 1 + count_digits(number // 10)

print(count_digits(987654))
````

### Explanation

Each recursive call removes one digit using integer division (`// 10`).

The recursion stops when only one digit remains.

# 🌍 Real-World Applications of Recursion

## 🌍 Example 1: Folder Traversal

Computers use recursion to explore folders and subfolders.

```text
Documents/
│
├── Python/
│   ├── Basics/
│   └── Projects/
│
└── Photos/
```

A recursive function can visit every folder and its subfolders automatically.

## 🌍 Example 2: Family Tree

A family tree has the same repeating structure.

```text
Grandparent
     │
Parent
     │
Child
```

Recursion can be used to move through each generation.

## 🌍 Example 3: Website Menu

Many websites have menus inside menus.

```text
Products
│
├── Electronics
│      ├── Mobile
│      └── Laptop
│
└── Clothing
```

Recursion helps display nested menus efficiently.

## 🌍 Example 4: Binary Search

Binary Search repeatedly divides the search space into smaller parts until the required element is found.

This divide-and-conquer approach is commonly implemented using recursion.

## 🌍 Example 5: Tree Traversal

Tree data structures are widely used in file systems, databases, and search engines.

Recursive functions naturally visit every node of a tree.

## 🔄 Recursion vs Loops

| Feature                                   | Recursion      | Loops          |
| ----------------------------------------- | -------------- | -------------- |
| Repeats code                              | ✅              | ✅              |
| Uses function calls                       | ✅              | ❌              |
| Needs a base condition                    | ✅              | ❌              |
| Easy for repetitive mathematical problems | ✅              | ⚠️ Sometimes   |
| Memory usage                              | Higher         | Lower          |
| Execution speed                           | Usually slower | Usually faster |


### When Should You Use Recursion?

Use recursion when:

- The problem can be divided into smaller versions of itself.
- Working with trees or graphs.
- Traversing folders and nested structures.
- Solving divide-and-conquer algorithms such as Binary Search or Merge Sort.

For simple repetitive tasks, loops are usually a better choice because they use less memory and are generally faster.

## ⚠️ RecursionError

````python
def infinite():
    infinite()

infinite()
````

output:-
 - RecursionError:
maximum recursion depth exceeded

### Explanation

The function keeps calling itself without a base case.

Python eventually stops the program and raises a **`RecursionError`** to prevent infinite recursion.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting to write a base case.
- ❌ Writing a base case that is never reached.
- ❌ Causing infinite recursion.
- ❌ Confusing recursion with loops.
- ❌ Using recursion for simple tasks where loops are more suitable.
- ❌ Forgetting to return the result of a recursive call.

## 💡 Best Practices

Follow these best practices while writing recursive functions:

- ✅ Always define a base case first.
- ✅ Ensure each recursive call moves closer to the base case.
- ✅ Keep recursive functions simple and easy to understand.
- ✅ Use meaningful function and parameter names.
- ✅ Prefer loops for simple repetitive tasks.
- ✅ Test recursive functions with small inputs before using larger ones.
