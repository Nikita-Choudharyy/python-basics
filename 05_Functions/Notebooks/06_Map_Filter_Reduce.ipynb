# 🗺️ Map, Filter, and Reduce in Python

## 📖 Introduction

Python provides powerful built-in tools for processing collections of data efficiently.

The three most commonly used functional programming tools are:

- 🗺️ `map()` – Transform data
- 🔍 `filter()` – Select data
- ➕ `reduce()` – Combine data into a single value

These functions are often used together with **lambda functions** to write clean and concise code.

# 🗺️ map()

## 🤔 What is `map()`?

The `map()` function applies a given function to **every element** of an iterable (such as a list or tuple).

It returns a **map object**, which is usually converted into a list using `list()`.

## 📝 Syntax

map(function, iterable)

### Syntax Explanation

- `function` → The function to apply.
- `iterable` → The collection of data (list, tuple, etc.).

## 💻 Example 1: Square Numbers

````python
numbers = [1, 2, 3, 4, 5]

result = list(map(lambda x: x ** 2, numbers))

print(result)
````

### Explanation

The lambda function squares each number in the list.

`map()` applies the function to every element and returns the transformed values.

## 💻 Example 2: Convert to Uppercase

````python
names = ["aman", "nikita", "riya"]

result = list(map(str.upper, names))

print(result)
````

### Explanation

The `str.upper` method is applied to each string in the list, converting all names to uppercase.

## 🌍 Real-World Example: Price with GST

````python
prices = [100, 250, 500]

gst_prices = list(map(lambda price: price * 1.18, prices))

print(gst_prices)
````

### Explanation

The GST is added to every product price using `map()`.

Each price is transformed into its updated value.

# 🔍 filter()

## 🤔 What is `filter()`?

The `filter()` function selects elements from an iterable based on a condition.

Only the elements for which the condition evaluates to **True** are included in the result.

Like `map()`, it returns a **filter object**, which is usually converted into a list using `list()`.

## 📝 Syntax

filter(function, iterable)


- `function` → A function that returns `True` or `False`.
- `iterable` → The collection of data to filter.

## 💻 Example 1: Even Numbers

````python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)
````

### Explanation

The lambda function checks whether each number is even.

Only the numbers that satisfy the condition are returned.

## 💻 Example 2: Odd Numbers

````python
numbers = [10, 15, 20, 25, 30, 35]

odd_numbers = list(filter(lambda x: x % 2 != 0, numbers))

print(odd_numbers)
````

### Explanation

The lambda function checks for odd numbers.

Only odd values are kept in the final list.

## 💻 Example 3: Numbers Greater Than 50

````python
numbers = [25, 60, 18, 90, 45, 72]

result = list(filter(lambda x: x > 50, numbers))

print(result)
````

### Explanation

Only the numbers greater than **50** satisfy the condition.

All other values are removed.

## 💻 Example 4: Filter Names by Length

````python
names = ["Aman", "Nikita", "Raj", "Priya", "An"]

result = list(filter(lambda name: len(name) >= 5, names))

print(result)
````

### Explanation

The condition keeps only names with **5 or more characters**.

## 🌍 Real-World Example: Passed Students

````python
marks = [35, 72, 48, 91, 60, 29]

passed_students = list(filter(lambda mark: mark >= 40, marks))

print(passed_students)
````

### Explanation

The function keeps only the students who scored **40 or above**.

Students with lower marks are filtered out.

## 🌍 Real-World Example: Available Products

````python
stock = [15, 0, 8, 20, 0, 12]

available_products = list(filter(lambda quantity: quantity > 0, stock))

print(available_products)
````

### Explanation

Products with a stock quantity greater than **0** are available for sale.

Out-of-stock items are removed.

## 🆚 `map()` vs `filter()`

| Feature            | `map()`               | `filter()`      |
| ------------------ | --------------------- | --------------- |
| Purpose            | Transform data        | Select data     |
| Output Size        | Usually same as input | Same or smaller |
| Condition Required | ❌ No                  | ✅ Yes           |
| Returns            | Modified values       | Filtered values |


### Example Comparison

````python
numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x ** 2, numbers))
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print("Squares:", squares)
print("Even Numbers:", even_numbers)
````

### Explanation

- `map()` transforms **every** element.
- `filter()` selects **only** the elements that satisfy a condition.

# ➕ reduce()

## 🤔 What is `reduce()`?

The `reduce()` function repeatedly applies a function to the elements of an iterable and combines them into a single result.

Unlike `map()` and `filter()`, which return iterables, `reduce()` returns a **single value**.

## 📦 Importing `reduce()`

````python
from functools import reduce
````

`reduce()` belongs to the **`functools`** module, so it must be imported before use.

## 📝 Syntax

reduce(function, iterable)

- `function` → A function that accepts **two arguments**.
- `iterable` → The collection of values to combine.

## 💻 Example 1: Sum of Numbers

````python
from functools import reduce

numbers = [10, 20, 30, 40]

total = reduce(lambda x, y: x + y, numbers)

print(total)
````

### Explanation

The lambda function adds two numbers at a time until only one final value remains.

```text
10 + 20 = 30
30 + 30 = 60
60 + 40 = 100
```

## 💻 Example 2: Product of Numbers

````python
from functools import reduce

numbers = [2, 3, 4, 5]

product = reduce(lambda x, y: x * y, numbers)

print(product)
````

### Explanation

The function multiplies two values at a time until a single result is produced.

```text
2 × 3 = 6
6 × 4 = 24
24 × 5 = 120
```

## 💻 Example 3: Find the Maximum Number

````python
from functools import reduce

numbers = [45, 12, 78, 23, 90, 56]

maximum = reduce(lambda x, y: x if x > y else y, numbers)

print(maximum)
````

### Explanation

The lambda function compares two numbers and keeps the larger one.

After comparing all values, the largest number remains.

## 🌍 Real-World Example: Total Sales

````python
from functools import reduce

sales = [1500, 2200, 1800, 2500]

total_sales = reduce(lambda x, y: x + y, sales)

print(total_sales)
````

### Explanation

The function combines all daily sales into one total sales value.

## 🌍 Real-World Example: Total Marks

````python
from functools import reduce

marks = [85, 90, 78, 88]

total_marks = reduce(lambda x, y: x + y, marks)

print(total_marks)
````

### Explanation

The function adds all subject marks to calculate the student's total score.

## 🆚 `map()` vs `filter()` vs `reduce()`

| Feature           | `map()`        | `filter()`    | `reduce()`   |
| ----------------- | -------------- | ------------- | ------------ |
| Purpose           | Transform data | Select data   | Combine data |
| Returns           | Iterable       | Iterable      | Single value |
| Uses Condition    | ❌              | ✅             | ❌            |
| Number of Results | Same as input  | Same or fewer | One          |


### Example Comparison

````python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

squares = list(map(lambda x: x ** 2, numbers))
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
total = reduce(lambda x, y: x + y, numbers)

print("Squares:", squares)
print("Even Numbers:", even_numbers)
print("Total:", total)
````

### Explanation

- 🗺️ **`map()`** transforms every element.
- 🔍 **`filter()`** keeps only elements that satisfy a condition.
- ➕ **`reduce()`** combines all elements into one final value.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Forgetting to convert the result of `map()` or `filter()` into a list using `list()`.
- ❌ Forgetting to import `reduce()` from the `functools` module.
- ❌ Using a lambda function that returns the wrong value.
- ❌ Confusing the purpose of `map()`, `filter()`, and `reduce()`.
- ❌ Using `reduce()` for simple operations where built-in functions like `sum()` or `max()` are more appropriate.

## 💡 Best Practices

Follow these best practices while using `map()`, `filter()`, and `reduce()`:

- ✅ Use `map()` when every element needs to be transformed.
- ✅ Use `filter()` when selecting elements based on a condition.
- ✅ Use `reduce()` when combining multiple values into a single result.
- ✅ Keep lambda functions short and readable.
- ✅ Use meaningful variable names.
- ✅ Choose the approach that makes your code easiest to understand.
