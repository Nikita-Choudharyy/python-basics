# 🌍 Variable Scope in Python

## 📖 Introduction

Variables are not always accessible from every part of a program.

The **scope** of a variable determines where it can be accessed and used.

Understanding variable scope helps you write cleaner programs and avoid unexpected errors.

## 🤔 What is Variable Scope?

Variable scope refers to the region of a program where a variable can be accessed.

Python mainly provides two types of variable scope:

- **Local Scope**
- **Global Scope**

## 🏠 Local Variables

A **local variable** is created inside a function.

It can only be accessed within that function.

Once the function finishes execution, the local variable is destroyed.

### Example 1

```python
def greet():
    message = "Hello!"

    print(message)

greet()
```

### Explanation

The variable `message` is created inside the function.

It is available only while the function is executing.

### Example 2

```python
def student():
    name = "Nikita"

    print(name)

student()

print(name)
```

### Explanation

The variable `name` exists only inside the `student()` function.

Trying to access it outside the function results in a **NameError**.

## 🌍 Global Variables

A **global variable** is created outside all functions.

It can be accessed both inside and outside functions.

### Example 1

```python
course = "Python"

def display():
    print(course)

display()

print(course)
```

### Explanation

The variable `course` is declared outside the function.

Since it is global, it can be accessed both inside and outside the function.

### Example 2

```python
city = "Jaipur"

def show_city():
    print(city)

show_city()
```

### Explanation

The function accesses the global variable directly because it is available throughout the program.

## 🌐 The `global` Keyword

Normally, a function can **read** a global variable.

However, if you want to **modify** a global variable inside a function, you must use the `global` keyword.

### Example 1: Reading a Global Variable

```python
message = "Welcome!"

def greet():
    print(message)

greet()
```

### Explanation

The function only reads the global variable.

Since the value is not modified, the `global` keyword is not required.

### Example 2: Modifying a Global Variable

```python
count = 0

def increase():
    global count
    count += 1

increase()

print(count)
```

### Explanation

The `global` keyword tells Python to use the global variable `count`.

Without `global`, Python would create a new local variable instead.

### Example 3: What Happens Without `global`?

```python
count = 0

def increase():
    count += 1

increase()
```

### Explanation

Python assumes `count` is a local variable because it is being assigned inside the function.

Since the local variable has no initial value, an `UnboundLocalError` occurs.

## ⏳ Lifetime of Variables

The **lifetime** of a variable is the period during which it exists in memory.

- **Local variables** exist only while the function is executing.
- **Global variables** exist until the program finishes.

### Example 1: Lifetime of a Local Variable

```python
def show():
    number = 100
    print(number)

show()
```

### Explanation

The variable `number` is created when the function starts.

It is automatically destroyed after the function finishes execution.

### Example 2: Lifetime of a Global Variable

```python
language = "Python"

def display():
    print(language)

display()

print(language)
```

### Explanation

The global variable exists throughout the execution of the program.

It can be accessed multiple times until the program ends.

## 🌍 Real-World Example 1: Website Visitor Counter

```python
visitors = 0

def visit():
    global visitors
    visitors += 1
    print(f"Visitors: {visitors}")

visit()
visit()
visit()
```

### Explanation

The global variable `visitors` stores the total number of website visitors.

Each function call updates the same global variable.

## 🌍 Real-World Example 2: Game Score

```python
score = 0

def add_score():
    global score
    score += 10
    print(f"Score: {score}")

add_score()
add_score()
```

### Explanation

The player's score is stored in a global variable.

Each time the player earns points, the function updates the same score.

## 🌍 Real-World Example 3: Bank Account Balance

```python
balance = 1000

def deposit(amount):
    global balance
    balance += amount
    print(f"Updated Balance: ₹{balance}")

deposit(500)
deposit(1000)
```

### Explanation

The account balance is stored as a global variable.

Whenever money is deposited, the function updates the balance.

## ⚠️ Common Mistakes

Beginners often make these mistakes:

- ❌ Trying to access a local variable outside its function.
- ❌ Confusing local and global variables.
- ❌ Forgetting to use the `global` keyword when modifying a global variable.
- ❌ Using too many global variables, making the program difficult to maintain.
- ❌ Giving local and global variables the same name without understanding variable scope.

## 💡 Best Practices

Follow these best practices while working with variable scope:

- ✅ Prefer local variables whenever possible.
- ✅ Use global variables only when multiple functions need shared data.
- ✅ Avoid modifying global variables unless necessary.
- ✅ Choose meaningful variable names.
- ✅ Keep the scope of variables as small as possible.
